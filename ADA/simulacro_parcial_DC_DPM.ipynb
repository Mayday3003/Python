### 🔢 Problema: Integer Replacement

Dado un número entero positivo `n`, se pueden aplicar las siguientes operaciones:

* Si `n` es **par**, reemplázalo por `n / 2`.
* Si `n` es **impar**, reemplázalo por `n + 1` o `n - 1`.

Tu tarea consiste en **retornar el número mínimo de operaciones necesarias para que `n` se convierta en 1**.

---

#### Ejemplos

* **Entrada:** `n = 8`
  **Salida:** `3`
  **Explicación:** 8 → 4 → 2 → 1

* **Entrada:** `n = 7`
  **Salida:** `4`
  **Explicación:**

  * 7 → 8 → 4 → 2 → 1
  * ó 7 → 6 → 3 → 2 → 1

* **Entrada:** `n = 4`
  **Salida:** `2`
  **Explicación:** 4 → 2 → 1

---

#### 📌 Restricciones

* 1 ≤ n ≤ 2³¹ - 1


---
Leetcode 397

````python
def integerReplacement(n: int,memo = {},x=None) -> int:
    
    if x == None:
        x = n

    if x == 1:
        return 0
    if x in memo:
        return memo[x]
    if x % 2 == 0:
        memo[x] = 1 + integerReplacement(n,memo,x // 2)
    else: 
        memo[x] = 1 + min(integerReplacement(n,memo,x-1),integerReplacement(n,memo,x+1))
    return(memo[x])



print(integerReplacement(8))  
print(integerReplacement(7))  
````

### 🌲 Problema: Convertir Arreglo Ordenado a Árbol Binario de Búsqueda Balanceado

Dado un arreglo de enteros `nums` **ordenado en orden ascendente**, conviértelo en un **árbol binario de búsqueda (BST) balanceado en altura**.

* Un **BST** cumple que, para cada nodo, todos los valores del subárbol izquierdo son menores y los del derecho son mayores.
* **Balanceado en altura** significa que, para todo nodo, la diferencia entre las alturas de sus subárboles izquierdo y derecho es como máximo **1**.

🎯 **Objetivo:**
Construir y retornar la **raíz** de un BST balanceado en altura que contenga exactamente los elementos de `nums`.

> 🔎 Nota: Pueden existir múltiples respuestas correctas; **cualquier BST balanceado** que respete el orden es válido.

---

#### Ejemplos

* **Entrada:** `nums = [-10, -3, 0, 5, 9]`
  **Salida (forma nivel-orden):** `[0, -3, 9, -10, null, 5]`
  **Explicación:** También se acepta `[0, -10, 5, null, -3, null, 9]`, entre otras variantes balanceadas.

image.png

* **Entrada:** `nums = [1, 3]`
  **Salida (nivel-orden):** `[3, 1]`
  **Explicación:** `[1, null, 3]` y `[3, 1]` son BSTs balanceados válidos.

---

#### 📌 Restricciones

* `1 ≤ nums.length ≤ 10^4`
* `-10^4 ≤ nums[i] ≤ 10^4`
* `nums` está **estrictamente** creciente (sin duplicados).

---
Leetcode 108

````python
class TreeNode:
    def __init__(self, root=0, left=None, right=None):
        self.root = root
        self.left = left
        self.right = right

def sortedArray(nums, left=0, right=None):
    if right is None:
        right = len(nums) - 1
        
    if left > right:
        return None
    
    mid = (left + right) // 2
    root = TreeNode(nums[mid])
    root.left = sortedArray(nums,left,mid-1)
    root.right = sortedArray(nums,mid+1,right)
    return root


def tree_to_list(root):
    if not root:
        return []
    result = []
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.root)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)

    return result

nums = [-10, -3, 0, 5, 9]
tree = sortedArray(nums)
print(tree_to_list(tree))
````

### 📝 Problema: Longest Substring with At Least K Repeating Characters

Dada una cadena `s` y un entero `k`, encuentra la **longitud de la subcadena más larga** de `s` tal que la **frecuencia de cada carácter** en dicha subcadena sea **mayor o igual que `k`**.

* Si no existe ninguna subcadena que cumpla la condición, retorna `0`.

---

#### Ejemplos

* **Entrada:** `s = "aaabb"`, `k = 3`
  **Salida:** `3`
  **Explicación:** La subcadena más larga válida es `"aaa"`, ya que `'a'` aparece 3 veces.

* **Entrada:** `s = "ababbc"`, `k = 2`
  **Salida:** `5`
  **Explicación:** La subcadena más larga válida es `"ababb"`, ya que `'a'` aparece 2 veces y `'b'` aparece 3 veces.

---

#### 📌 Restricciones

* 1 ≤ s.length ≤ 10⁴
* `s` consiste solo en letras minúsculas del alfabeto inglés.
* 1 ≤ k ≤ 10⁵
