### ðŸ”¢ Problema: Integer Replacement

Dado un nÃºmero entero positivo `n`, se pueden aplicar las siguientes operaciones:

* Si `n` es **par**, reemplÃ¡zalo por `n / 2`.
* Si `n` es **impar**, reemplÃ¡zalo por `n + 1` o `n - 1`.

Tu tarea consiste en **retornar el nÃºmero mÃ­nimo de operaciones necesarias para que `n` se convierta en 1**.

---

#### Ejemplos

* **Entrada:** `n = 8`
  **Salida:** `3`
  **ExplicaciÃ³n:** 8 â†’ 4 â†’ 2 â†’ 1

* **Entrada:** `n = 7`
  **Salida:** `4`
  **ExplicaciÃ³n:**

  * 7 â†’ 8 â†’ 4 â†’ 2 â†’ 1
  * Ã³ 7 â†’ 6 â†’ 3 â†’ 2 â†’ 1

* **Entrada:** `n = 4`
  **Salida:** `2`
  **ExplicaciÃ³n:** 4 â†’ 2 â†’ 1

---

#### ðŸ“Œ Restricciones

* 1 â‰¤ n â‰¤ 2Â³Â¹ - 1


---
Leetcode 397

````python
def integerReplacement(n: int,memo = {},x=None) -> int:
    
    if x == None:
        x = n

    if x == 1:
        return 0
    if x in memo:
        return memo[x]
    if x % 2 == 0:
        memo[x] = 1 + integerReplacement(n,memo,x // 2)
    else: 
        memo[x] = 1 + min(integerReplacement(n,memo,x-1),integerReplacement(n,memo,x+1))
    return(memo[x])



print(integerReplacement(8))  
print(integerReplacement(7))  
````

### ðŸŒ² Problema: Convertir Arreglo Ordenado a Ãrbol Binario de BÃºsqueda Balanceado

Dado un arreglo de enteros `nums` **ordenado en orden ascendente**, conviÃ©rtelo en un **Ã¡rbol binario de bÃºsqueda (BST) balanceado en altura**.

* Un **BST** cumple que, para cada nodo, todos los valores del subÃ¡rbol izquierdo son menores y los del derecho son mayores.
* **Balanceado en altura** significa que, para todo nodo, la diferencia entre las alturas de sus subÃ¡rboles izquierdo y derecho es como mÃ¡ximo **1**.

ðŸŽ¯ **Objetivo:**
Construir y retornar la **raÃ­z** de un BST balanceado en altura que contenga exactamente los elementos de `nums`.

> ðŸ”Ž Nota: Pueden existir mÃºltiples respuestas correctas; **cualquier BST balanceado** que respete el orden es vÃ¡lido.

---

#### Ejemplos

* **Entrada:** `nums = [-10, -3, 0, 5, 9]`
  **Salida (forma nivel-orden):** `[0, -3, 9, -10, null, 5]`
  **ExplicaciÃ³n:** TambiÃ©n se acepta `[0, -10, 5, null, -3, null, 9]`, entre otras variantes balanceadas.

image.png

* **Entrada:** `nums = [1, 3]`
  **Salida (nivel-orden):** `[3, 1]`
  **ExplicaciÃ³n:** `[1, null, 3]` y `[3, 1]` son BSTs balanceados vÃ¡lidos.

---

#### ðŸ“Œ Restricciones

* `1 â‰¤ nums.length â‰¤ 10^4`
* `-10^4 â‰¤ nums[i] â‰¤ 10^4`
* `nums` estÃ¡ **estrictamente** creciente (sin duplicados).

---
Leetcode 108

````python
class TreeNode:
    def __init__(self, root=0, left=None, right=None):
        self.root = root
        self.left = left
        self.right = right

def sortedArray(nums, left=0, right=None):
    if right is None:
        right = len(nums) - 1
        
    if left > right:
        return None
    
    mid = (left + right) // 2
    root = TreeNode(nums[mid])
    root.left = sortedArray(nums,left,mid-1)
    root.right = sortedArray(nums,mid+1,right)
    return root


def tree_to_list(root):
    if not root:
        return []
    result = []
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.root)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)

    return result

nums = [-10, -3, 0, 5, 9]
tree = sortedArray(nums)
print(tree_to_list(tree))
````
