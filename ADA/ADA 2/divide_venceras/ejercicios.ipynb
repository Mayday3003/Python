```python
"""
═══════════════════════════════════════════════════════════════════════════════
        DIVIDE Y VENCERÁS - EJERCICIOS EXTRA (FÁCIL-INTERMEDIO)
═══════════════════════════════════════════════════════════════════════════════
"""

# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 6: Encontrar el Mínimo en un Arreglo Rotado
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Un arreglo ordenado fue rotado k posiciones. Encontrar el mínimo.
#
# Ejemplo:
#   [4,5,6,7,1,2,3] → 1
#   [1,2,3,4,5]     → 1
#   [2,1]           → 1
#    

# D&V: O(log n)

   
def minimo_rotado(arr, izq = 0, der = None):
    
    if der is None:
        der = len(arr) -1
        
    if len(arr) == 1:
        return arr[0]
        
        
    if arr[izq] <= arr[der]:
        return arr[izq]
    

    
    mid = (izq + der) // 2

    if arr[mid] >= arr[der]:
        return minimo_rotado(arr, mid + 1, der)
    else:
        explorar_izq = minimo_rotado(arr, izq, mid -1)
        return min(arr[mid], explorar_izq)
        



print(minimo_rotado([3, 1, 2]))



# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 7: Contar Ocurrencias de x en Arreglo Ordenado
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Dado arr ordenado y x, contar cuántas veces aparece x.
#
# Ejemplo:
#   arr=[1,2,2,2,3,4], x=2 → 3
#   arr=[1,1,1,1], x=1      → 4
#   arr=[1,2,3], x=5        → 0
#
# D&V: O(log n)

def repeticiones_numero(arr: list, x: int, izq = 0, der =  None):
    if der is None:
        der = len(arr) - 1
    if izq > der:
        return 0

    mid = (izq + der) // 2

    # tu misma lógica, en versión larga:
    if arr[mid] == x:
        cuenta_mid = 1
    else:
        cuenta_mid = 0

    izquierda = repeticiones_numero(arr, x, izq, mid - 1)
    derecha = repeticiones_numero(arr, x, mid + 1, der)

    return cuenta_mid + izquierda + derecha


repeticiones_numero([2,1,2,2,2],2)
    



# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 8: Encontrar un Pico en un Arreglo
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Encontrar un índice i tal que arr[i] >= vecinos.
#
# Ejemplo:
#   [1,3,20,4,1,0] → índice 2 (valor 20)
#   [10,20,15,2,23,90,67] → puede ser 1 o 5
#
# D&V: O(log n)

def encontrar_pico_solucion(arr):
    if not arr:
        return None

    def dyv(lo, hi):
        if lo == hi:
            return lo

        mid = (lo + hi) // 2
        if arr[mid] < arr[mid + 1]:
            return dyv(mid + 1, hi)
        return dyv(lo, mid)

    idx = dyv(0, len(arr) - 1)
    return idx, arr[idx]


# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 9: Potencia Rápida (x^n)
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Calcular x^n usando D&V.
#
# Ejemplo:
#   x=2, n=10 → 1024
#   x=3, n=5  → 243
#   x=2, n=0  → 1
#
# D&V: O(log n)

def potencia_rapida_solucion(x, n):
    if n == 0:
        return 1
    if n < 0:
        return 1 / potencia_rapida_solucion(x, -n)

    mitad = potencia_rapida_solucion(x, n // 2)
    if n % 2 == 0:
        return mitad * mitad
    return mitad * mitad * x


# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 10: Elemento Mayoritario (LeetCode 169)
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Encontrar el elemento que aparece más de n/2 veces.
#
# Ejemplo:
#   [3,2,3]       → 3
#   [2,2,1,1,1,2,2] → 2
#
# D&V: O(n log n)

def elemento_mayoritario_solucion(arr):
    if not arr:
        return None

    def dyv(lo, hi):
        if lo == hi:
            return arr[lo]

        mid = (lo + hi) // 2
        izq = dyv(lo, mid)
        der = dyv(mid + 1, hi)

        if izq == der:
            return izq

        c_izq = 0
        c_der = 0
        for i in range(lo, hi + 1):
            if arr[i] == izq:
                c_izq += 1
            elif arr[i] == der:
                c_der += 1

        return izq if c_izq > c_der else der

    return dyv(0, len(arr) - 1)
```

```python
"""
═══════════════════════════════════════════════════════════════════════════════
      DIVIDE Y VENCERÁS - EJERCICIOS DIFERENTES (FÁCIL-INTERMEDIO)
═══════════════════════════════════════════════════════════════════════════════
"""

# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 11: Contar Inversiones en un Arreglo
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Contar cuántos pares (i, j) existen con i < j y arr[i] > arr[j].
#
# Ejemplo:
#   [2,4,1,3,5] → 3
#   [1,2,3]     → 0
#
# D&V: O(n log n)

def contar_inversiones(arr):
    # TU CÓDIGO AQUÍ
    pass


def contar_inversiones_solucion(arr):
    def dyv(nums):
        if len(nums) <= 1:
            return nums, 0

        mid = len(nums) // 2
        izq, inv_izq = dyv(nums[:mid])
        der, inv_der = dyv(nums[mid:])

        merged = []
        i = j = 0
        inv_split = 0

        while i < len(izq) and j < len(der):
            if izq[i] <= der[j]:
                merged.append(izq[i])
                i += 1
            else:
                merged.append(der[j])
                inv_split += len(izq) - i
                j += 1

        merged.extend(izq[i:])
        merged.extend(der[j:])
        return merged, inv_izq + inv_der + inv_split

    return dyv(arr)[1]


# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 12: Máxima Suma de Subarreglo (Maximum Subarray)
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Encontrar la suma máxima de un subarreglo contiguo.
#
# Ejemplo:
#   [-2,1,-3,4,-1,2,1,-5,4] → 6
#   [1]                      → 1
#
# D&V: O(n log n)

def max_subarreglo(arr):
    # TU CÓDIGO AQUÍ
    pass


def max_subarreglo_solucion(arr):
    if not arr:
        return None

    def dyv(lo, hi):
        if lo == hi:
            return arr[lo]

        mid = (lo + hi) // 2
        best_izq = dyv(lo, mid)
        best_der = dyv(mid + 1, hi)

        suma = 0
        mejor_sufijo_izq = float("-inf")
        for i in range(mid, lo - 1, -1):
            suma += arr[i]
            mejor_sufijo_izq = max(mejor_sufijo_izq, suma)

        suma = 0
        mejor_prefijo_der = float("-inf")
        for i in range(mid + 1, hi + 1):
            suma += arr[i]
            mejor_prefijo_der = max(mejor_prefijo_der, suma)

        cruzado = mejor_sufijo_izq + mejor_prefijo_der
        return max(best_izq, best_der, cruzado)

    return dyv(0, len(arr) - 1)


# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 13: Multiplicación Rápida de Enteros (Karatsuba)
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Multiplicar dos enteros grandes usando D&V.
#
# Ejemplo:
#   1234 * 5678 → 7006652
#
# D&V: ~O(n^1.585)

def karatsuba(x, y):
    # TU CÓDIGO AQUÍ
    pass


def karatsuba_solucion(x, y):
    signo = -1 if (x < 0) ^ (y < 0) else 1
    x, y = abs(x), abs(y)

    def dyv(a, b):
        if a < 10 or b < 10:
            return a * b

        n = max(len(str(a)), len(str(b)))
        m = n // 2
        p10 = 10 ** m

        a_hi, a_lo = divmod(a, p10)
        b_hi, b_lo = divmod(b, p10)

        z0 = dyv(a_lo, b_lo)
        z2 = dyv(a_hi, b_hi)
        z1 = dyv(a_lo + a_hi, b_lo + b_hi)

        return z2 * (10 ** (2 * m)) + (z1 - z2 - z0) * p10 + z0

    return signo * dyv(x, y)


# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 14: Encontrar Punto Fijo en Arreglo Ordenado
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Hallar un índice i tal que arr[i] == i (si existe).
#
# Ejemplo:
#   [-10,-5,0,3,7] → 3
#   [0,2,5,8,17]   → 0
#   [-10,-5,3,4,7] → None
#
# D&V: O(log n)

def punto_fijo(arr):
    # TU CÓDIGO AQUÍ
    pass


def punto_fijo_solucion(arr):
    def dyv(lo, hi):
        if lo > hi:
            return None
        mid = (lo + hi) // 2

        if arr[mid] == mid:
            return mid
        if arr[mid] < mid:
            return dyv(mid + 1, hi)
        return dyv(lo, mid - 1)

    return dyv(0, len(arr) - 1)


# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 15: Distancia Mínima entre Dos Valores en Arreglo
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Dado un arreglo, encontrar la mínima diferencia absoluta entre
# dos elementos.
#
# Ejemplo:
#   [8,1,5,12,10] → 2   (10 y 12, o 5 y 8)
#
# Idea D&V:
#   - Ordenar con merge sort (D&V)
#   - Buscar mínima diferencia entre vecinos
#
# D&V: O(n log n)

def minima_diferencia_solucion(arr):
    if len(arr) < 2:
        return None

    def merge_sort(nums):
        if len(nums) <= 1:
            return nums
        m = len(nums) // 2
        izq = merge_sort(nums[:m])
        der = merge_sort(nums[m:])

        out = []
        i = j = 0
        while i < len(izq) and j < len(der):
            if izq[i] <= der[j]:
                out.append(izq[i]); i += 1
            else:
                out.append(der[j]); j += 1
        out.extend(izq[i:])
        out.extend(der[j:])
        return out

    ordenado = merge_sort(arr)
    mejor = float("inf")
    for i in range(1, len(ordenado)):
        mejor = min(mejor, ordenado[i] - ordenado[i - 1])
    return mejor
```
