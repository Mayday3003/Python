```python
# EJERCICIO 1: Elemento mayoritario
# ─────────────────────────────────────────────────────────────────────────────
# Dado un arreglo de enteros, encontrar el elemento que aparece más de n/2 veces.
# Se garantiza que siempre existe un elemento mayoritario.
#
# Ejemplo:
#   entrada: [3, 3, 4, 2, 3, 3, 5, 3, 3]
#   salida: 3 (aparece 6 veces, n/2 = 4.5)
#
def elemento_mayoritario(arr:list[int])->int:
    cantidad_elementos = {}

    for i in arr:
        if i not in cantidad_elementos:

            cantidad_elementos[i] = 1
        else: 
            cantidad_elementos[i] +=1
        print(cantidad_elementos)
        
    for clave, valor in cantidad_elementos.items():
     if valor >= (len(arr)//2):
        return clave

elemento_mayoritario([3, 3, 4, 2, 3, 3, 5, 3, 3])

```

```python

# EJERCICIO 2: Subarreglo de suma máxima
# ─────────────────────────────────────────────────────────────────────────────
# Dado un arreglo de enteros (puede tener negativos), encontrar el subarreglo
# contiguo con la suma más grande. Retornar la suma.
#
# Ejemplo:
#   entrada: [-2, 1, -3, 4, -1, 2, 1, -5, 4]
#   salida: 6 (el subarreglo [4, -1, 2, 1] tiene suma 6)
#
def suma_maxima_subarreglo(arr):
    max_sum = float('-inf')
    n = len(arr)
   
    for i in range(n):
        for j in range(i,n):

            current_max_sum = 0 
            for k in range(i,j+1):
                current_max_sum += arr[k] 
            max_sum = max(max_sum,current_max_sum)        
    return max_sum

suma_maxima_subarreglo([-2, 1, -3, 4, -1, 2, 1, -5, 4])
```

```python

# EJERCICIO 2: Subarreglo de suma máxima
# ─────────────────────────────────────────────────────────────────────────────
# Dado un arreglo de enteros (puede tener negativos), encontrar el subarreglo
# contiguo con la suma más grande. Retornar la suma.
#
# Ejemplo:
#   entrada: [-2, 1, -3, 4, -1, 2, 1, -5, 4]
#   salida: 6 (el subarreglo [4, -1, 2, 1] tiene suma 6)
#
def suma_maxima_subarreglo(arr):
    max_sum = float('-inf')
   
    for i in range(len(arr)):
        current_max_sum = 0 

        for j in range(i,(len(arr)-1)):
            current_max_sum += arr[j] 
            if current_max_sum > max_sum:
                max_sum = current_max_sum
    return max_sum

suma_maxima_subarreglo([-2, 1, -3, 4, -1, 2, 1, -5, 4])
```

```python
# EJERCICIO 3: Tres elementos que sumen cero
# ─────────────────────────────────────────────────────────────────────────────
# Dado un arreglo de enteros, determinar si existen tres elementos distintos
# (por posición) cuya suma sea exactamente 0. Retornar True/False.
#
# Ejemplo:
#   entrada: [-1, 0, 1, 2, -1, -4]
#   salida: True (porque -1 + 0 + 1 = 0)
#
def tres_suman_cero(arr):
    n = len(arr)
    for i in range(n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                respuesta = arr[i] + arr[j] + arr[k] 
                if respuesta == 0:
                    return True
    return False


tres_suman_cero([-1, 0, 50, 2, -20, -4])
```

```python
# EJERCICIO 4: Cadena con todos los caracteres únicos
# ─────────────────────────────────────────────────────────────────────────────
# Dada una cadena, determinar si todos sus caracteres son únicos
# (no se repite ninguno). NO usar estructuras de datos adicionales.
#
# Ejemplo:
#   entrada: "abcdef"  → True
#   entrada: "abcaef"  → False (la 'a' se repite)
#
def caracteres_unicos(texto):
    caracteres = set()
    for i in texto:
        if i in caracteres:
            return False
        caracteres.add(i)
    return True
    

caracteres_unicos("abcaef")
```

```python
# EJERCICIO 5: Potencia de un número (recursivo)
# ─────────────────────────────────────────────────────────────────────────────
# Calcular base^exponente usando recursión por fuerza bruta.
# No usar el operador ** ni la función pow().
#
# Ejemplo:
#   entrada: base=2, exponente=10
#   salida: 1024
#
def calculo_potencia(base, exponente,potencia = 0):
    if exponente == 1:
        return potencia


    return calculo_potencia(base, exponente-1,base*base)

calculo_potencia(2,10)
```

```python
# Recursivo
def is_palindromo(palabra:str, i = 0, j = None)->bool:
    if j == None:
        j = len(palabra)-1

    if palabra[i] != palabra[j]:
        return False
    if j == i:
        return True

    return is_palindromo(palabra, i+1,j-1)

is_palindromo('reconocer')
is_palindromo('oeovbn')
```
