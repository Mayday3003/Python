````python
def max_sum_3(n: int, arr:list[int])-> int:
    max_sum = float('-inf')

    for i in range(n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                current_sum = arr[i] + arr[j] + arr[k]
                if current_sum >= max_sum:
                    max_sum = current_sum
    return max_sum

max_sum_3(5,[1,2,3,4,5])
````

$
[

]
$

Si no existe, imprime `-1`.

Los extremos (posición 0 y posición n−1) **no pueden ser pico**.

* Un entero ( n ) (1 ≤ n ≤ 2e5)

* O `-1` si no existe

5

Salida:

1
```

```

Salida:
