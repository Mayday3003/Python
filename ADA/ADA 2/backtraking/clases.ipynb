```python
# Generar todas las cadenas binarias de longitud n sin 1s consecutivos

def cadenas_binarias_sin_unos_consecutivos(n):
    resultado = []

    def backtrack(pos, cadena_actual):
        # Caso base: ya construimos n posiciones
        if pos == n:
            resultado.append("".join(cadena_actual))
            return

        # Opcion 1: colocar 0 siempre es valido
        cadena_actual.append("0")
        backtrack(pos + 1, cadena_actual)
        cadena_actual.pop()

        # Opcion 2: colocar 1 solo si el anterior no fue 1
        if not cadena_actual or cadena_actual[-1] != "1":
            cadena_actual.append("1")
            backtrack(pos + 1, cadena_actual)
            cadena_actual.pop()

    backtrack(0, [])
    return resultado

# Ejemplo
n = 3
cadenas = cadenas_binarias_sin_unos_consecutivos(n)
print(cadenas)
print("Cantidad:", len(cadenas))
```

## Explicacion del algoritmo (Backtracking)

Queremos construir cadenas binarias de longitud `n` cumpliendo una restriccion: **no puede haber `11` consecutivo**.

La idea de backtracking es construir la cadena posicion por posicion:

1. Elegimos que poner en la posicion actual (`0` o `1`).
2. Exploramos recursivamente esa decision.
3. Nos devolvemos (hacemos `pop`) para probar otra opcion.

### Regla de poda

- `0` se puede poner siempre.
- `1` solo se puede poner si la posicion anterior **no** tiene `1`.

Esa validacion evita generar caminos invalidos, por eso es eficiente.

### Caso base

Cuando `pos == n`, significa que ya completamos una cadena valida y la guardamos en `resultado`.

### Ejemplo para `n = 3`

Una salida valida puede verse asi (el orden puede variar segun el recorrido):

`['000', '001', '010', '100', '101']`

### Complejidad

- Tiempo: proporcional al numero de soluciones generadas (aprox. crecimiento tipo Fibonacci).
- Espacio: `O(n)` por la pila de recursion y la cadena en construccion.

```python
def cadenas_ab_hasta_dos_consecutivos(n):
    resultado = []

    def backtrack(cadena_actual):
        # Caso base: ya construimos una cadena de longitud n
        if len(cadena_actual) == n:
            resultado.append("".join(cadena_actual))
            return

        # Probar decisiones: agregar 'a' o agregar 'b'
        for letra in ["a", "b"]:
            # Poda: evitar tres letras iguales consecutivas (aaa o bbb)
            if len(cadena_actual) >= 2 and cadena_actual[-1] == letra and cadena_actual[-2] == letra:
                continue

            cadena_actual.append(letra)   # Elegir
            backtrack(cadena_actual)       # Explorar
            cadena_actual.pop()            # Devolverse

    backtrack([])
    return resultado

# Ejemplo
n = 3
cadenas = cadenas_ab_hasta_dos_consecutivos(n)
print(cadenas)
print("Cantidad:", len(cadenas))
```
