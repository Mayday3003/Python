```python
def coin_change_greedy(cambio:int, monedas: list[int]) -> int:
    monedas.sort(reverse=True)
    resultado = []
    for moneda in monedas:
        while cambio >= moneda:
            cambio -= moneda
            resultado.append(moneda)
    return resultado

# Ejemplo
print(coin_change_greedy(63, [1, 5, 10, 25]))
# Output: [25, 25, 10, 1, 1, 1]
```

```python
# Mayor cantidad de eventos 
eventos = [(1,4),(3,5),(0,6),(5,7),(6,10),(8,11)]

def seleccion_eventos(eventos):
    if not eventos:
        return None

    eventos = sorted(eventos, key =lambda x: x[1])
    programados = [eventos[0]]
    ultimo_fin = eventos[0][1]

    for inicio, fin in eventos[1:]:
        if inicio >= ultimo_fin:
            programados.append((inicio,fin))
            ultimo_fin = fin

seleccion_eventos(eventos)
```

```python
def mochila_fraccionaria(capacidad, objetos):

    objetos_con_ratio = [(p,v,v/p) for p,v in objetos]
    objetos_con_ratio.sort(key= lambda x: x[2], reverse=True)

        
    valor_total = 0
    objetos_almacenados = []

    for peso,valor,ratio in objetos_con_ratio:
        if capacidad >= peso:
            valor_total += valor
            capacidad -= peso
            objetos_almacenados.append((peso,valor,1))
        elif capacidad > 0:
            fraccion = capacidad/peso
            valor_total += valor*fraccion
            objetos_almacenados.append(peso,valor,fraccion)
            break
    return objetos_almacenados, valor_total


mochila_fraccionaria(100,[(20,40),(10,30)])
```

```python
#minimo de paradas necesarias para llegar al destino GREEDY
def menor_paradas_posibles(distancia_total:int, capacidad_carro: int,
                            estaciones: list[int])-> list[int]:
    estaciones = [0] + sorted(estaciones) + [distancia_total]

    paradas = []

    combustible = capacidad_carro

    for i in range(1, len(estaciones)):
        distancia = estaciones[i] - estaciones[i-1]

        if distancia > capacidad_carro:
            return -1

        if distancia > combustible:
            paradas.append(estaciones[i-1])
            combustible = capacidad_carro

        combustible -= distancia


    return paradas



menor_paradas_posibles(50, 20, [15,20,40,50])
```

```python
#minimo de paradas necesarias para llegar al destino NO GREEDY
def menor_paradas_posibles(distancia_total:int, capacidad_carro: int,
                            estaciones: list[int])-> list[int]:

    mayor_recorrido = [(d, capacidad_carro - d) for d in estaciones]

    for parada, distancia in mayor_recorrido:
        if distancia < 0:
            distancia = float('inf')
        

    return (mayor_recorrido)

menor_paradas_posibles(50, 20, [15,20,40,50])
```
