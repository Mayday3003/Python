```python
"""
═══════════════════════════════════════════════════════════════════════════════
        PROGRAMACIÓN DINÁMICA — INTRODUCCIÓN Y MEMOIZATION (TOP-DOWN)
═══════════════════════════════════════════════════════════════════════════════

Programación Dinámica (DP) resuelve problemas que tienen:
    1. Subestructura óptima: la solución óptima se construye con
       soluciones óptimas de subproblemas
    2. Subproblemas solapados: los mismos subproblemas se resuelven
       muchas veces

Enfoque Top-Down (Memoization):
    - Resolver recursivamente (como siempre)
    - Guardar cada resultado en un diccionario/tabla
    - Antes de calcular, revisar si ya lo calculaste

    Si ya lo calculé → lo retorno directo
    Si no → lo calculo, lo guardo, y lo retorno
"""

import time


# ═══════════════════════════════════════════════════════════════════════════
# EJEMPLO 1: Fibonacci
# ═══════════════════════════════════════════════════════════════════════════
#
# F(0) = 0, F(1) = 1
# F(n) = F(n-1) + F(n-2)
#
# Sin DP: O(2^n) — recalcula los mismos valores muchas veces
# Con DP: O(n) — cada valor se calcula una sola vez
#

# --- RECURSIÓN PURA (sin DP) ---
def fib_recursivo(n):
    """Fibonacci recursivo. O(2^n) — MUY lento para n grande."""
    if n <= 1:
        return n
    return fib_recursivo(n - 1) + fib_recursivo(n - 2)


# --- CON MEMOIZATION (Top-Down DP) ---
def fib_memo(n, memo=None):
    """Fibonacci con memoization. O(n)."""
    if memo is None:
        memo = {}

    if n <= 1:
        return n

    # ¿Ya lo calculé?
    if n in memo:
        return memo[n]

    # Si no, lo calculo y lo guardo
    memo[n] = fib_memo(n - 1, memo) + fib_memo(n - 2, memo)
    return memo[n]


# ═══════════════════════════════════════════════════════════════════════════
# EJEMPLO 2: Subir Escaleras
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Tienes una escalera de n escalones. En cada paso puedes
#           subir 1 o 2 escalones. ¿De cuántas formas puedes llegar
#           al escalón n?
#
# Ejemplo:
#   n=1 → 1 forma:  (1)
#   n=2 → 2 formas: (1+1), (2)
#   n=3 → 3 formas: (1+1+1), (1+2), (2+1)
#   n=4 → 5 formas: (1+1+1+1), (1+1+2), (1+2+1), (2+1+1), (2+2)
#
# Recurrencia: formas(n) = formas(n-1) + formas(n-2)
#   Desde el escalón n, pudiste haber llegado desde n-1 (un paso)
#   o desde n-2 (dos pasos).
#
# ¡Es Fibonacci disfrazado!
#

# --- RECURSIÓN PURA ---
def escaleras_recursivo(n):
    """Formas de subir n escalones. O(2^n)."""
    if n <= 1:
        return 1
    return escaleras_recursivo(n - 1) + escaleras_recursivo(n - 2)


# --- CON MEMOIZATION ---
def escaleras_memo(n, memo=None):
    """Formas de subir n escalones con memoization. O(n)."""
    if memo is None:
        memo = {}

    if n <= 1:
        return 1

    if n in memo:
        return memo[n]

    memo[n] = escaleras_memo(n - 1, memo) + escaleras_memo(n - 2, memo)
    return memo[n]


# ═══════════════════════════════════════════════════════════════════════════
# EJEMPLO 3: Cambio de Monedas (Mínimo número de monedas)
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Dadas monedas de ciertas denominaciones y un monto objetivo,
#           encontrar el mínimo número de monedas para formar ese monto.
#           Hay monedas ilimitadas de cada tipo.
#
# Ejemplo:
#   monedas = [1, 5, 10], monto = 12
#   → 3 monedas (10 + 1 + 1)
#
#   monedas = [1, 3, 4], monto = 6
#   → 2 monedas (3 + 3)  — NO es greedy (greedy daría 4+1+1 = 3 monedas)
#
# Recurrencia:
#   minMonedas(monto) = 1 + min(minMonedas(monto - moneda))
#                           para cada moneda que quepa
#
# Sin DP: O(k^n) donde k = número de denominaciones
# Con DP: O(n × k)
#

# --- RECURSIÓN PURA ---
def monedas_recursivo(monedas, monto):
    """Mínimo de monedas para el monto. Exponencial."""
    if monto == 0:
        return 0
    if monto < 0:
        return float('inf')

    minimo = float('inf')
    for moneda in monedas:
        resultado = monedas_recursivo(monedas, monto - moneda)
        minimo = min(minimo, resultado + 1)

    return minimo


# --- CON MEMOIZATION ---
def monedas_memo(monedas, monto, memo=None):
    """Mínimo de monedas con memoization. O(monto × len(monedas))."""
    if memo is None:
        memo = {}

    if monto == 0:
        return 0
    if monto < 0:
        return float('inf')

    if monto in memo:
        return memo[monto]

    minimo = float('inf')
    for moneda in monedas:
        resultado = monedas_memo(monedas, monto - moneda, memo)
        minimo = min(minimo, resultado + 1)

    memo[monto] = minimo
    return minimo


# ═══════════════════════════════════════════════════════════════════════════
# EJEMPLO 4: Mochila 0/1 (Knapsack)
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Tienes n objetos, cada uno con un peso y un valor.
#           Tu mochila soporta un peso máximo W.
#           Maximizar el valor total sin exceder el peso.
#           Cada objeto se toma completo o no se toma (0/1).
#
# Ejemplo:
#   pesos  = [2, 3, 4, 5]
#   valores = [3, 4, 5, 6]
#   capacidad = 8
#   → Mejor: tomar objetos de peso 3 y 5 → valor = 4 + 6 = 10
#
# Recurrencia:
#   mochila(i, cap) = max(
#       mochila(i+1, cap),                    # NO tomar objeto i
#       valores[i] + mochila(i+1, cap-pesos[i])  # SÍ tomar objeto i
#   )
#
# Sin DP: O(2^n) — cada objeto se toma o no
# Con DP: O(n × W)
#

# --- RECURSIÓN PURA ---
def mochila_recursivo(pesos, valores, capacidad, i=0):
    """Mochila 0/1 recursiva. O(2^n)."""
    if i >= len(pesos) or capacidad <= 0:
        return 0

    # Opción 1: no tomar el objeto i
    sin_tomar = mochila_recursivo(pesos, valores, capacidad, i + 1)

    # Opción 2: tomar el objeto i (si cabe)
    con_tomar = 0
    if pesos[i] <= capacidad:
        con_tomar = valores[i] + mochila_recursivo(
            pesos, valores, capacidad - pesos[i], i + 1
        )

    return max(sin_tomar, con_tomar)


# --- CON MEMOIZATION ---
def mochila_memo(pesos, valores, capacidad, i=0, memo=None):
    """Mochila 0/1 con memoization. O(n × capacidad)."""
    if memo is None:
        memo = {}

    if i >= len(pesos) or capacidad <= 0:
        return 0

    # La clave del memo es (índice del objeto, capacidad restante)
    clave = (i, capacidad)
    if clave in memo:
        return memo[clave]

    sin_tomar = mochila_memo(pesos, valores, capacidad, i + 1, memo)

    con_tomar = 0
    if pesos[i] <= capacidad:
        con_tomar = valores[i] + mochila_memo(
            pesos, valores, capacidad - pesos[i], i + 1, memo
        )

    memo[clave] = max(sin_tomar, con_tomar)
    return memo[clave]



# ═══════════════════════════════════════════════════════════════════════════
# EJEMPLO 5: Subsecuencia Común Más Larga (LCS)
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Dadas dos cadenas, encontrar la longitud de la subsecuencia
#           común más larga. Una subsecuencia no necesita ser contigua.
#
# Ejemplo:
#   "ABCBDAB" y "BDCAB" → 4 ("BCAB")
#   "GATO" y "GRITO"    → 3 ("GTO")
#
# Recurrencia:
#   Si s1[i] == s2[j]: LCS(i,j) = 1 + LCS(i+1, j+1)
#   Si no:             LCS(i,j) = max(LCS(i+1, j), LCS(i, j+1))
#
# Sin DP: O(2^(n+m))
# Con DP: O(n × m)
#

# --- RECURSIÓN PURA ---
def lcs_recursivo(s1, s2, i=0, j=0):
    """LCS recursivo. Exponencial."""
    if i >= len(s1) or j >= len(s2):
        return 0

    if s1[i] == s2[j]:
        return 1 + lcs_recursivo(s1, s2, i + 1, j + 1)
    else:
        return max(
            lcs_recursivo(s1, s2, i + 1, j),
            lcs_recursivo(s1, s2, i, j + 1)
        )


# --- CON MEMOIZATION ---
def lcs_memo(s1, s2, i=0, j=0, memo=None):
    """LCS con memoization. O(n × m)."""
    if memo is None:
        memo = {}

    if i >= len(s1) or j >= len(s2):
        return 0

    clave = (i, j)
    if clave in memo:
        return memo[clave]

    if s1[i] == s2[j]:
        memo[clave] = 1 + lcs_memo(s1, s2, i + 1, j + 1, memo)
    else:
        memo[clave] = max(
            lcs_memo(s1, s2, i + 1, j, memo),
            lcs_memo(s1, s2, i, j + 1, memo)
        )

    return memo[clave]


# ═══════════════════════════════════════════════════════════════════════════
#                           DEMOSTRACIÓN
# ═══════════════════════════════════════════════════════════════════════════

if __name__ == "__main__":
    print("=" * 70)
    print("PROGRAMACIÓN DINÁMICA — MEMOIZATION (TOP-DOWN)")
    print("=" * 70)

    # ── Ejemplo 1: Fibonacci ──
    print("\n1. FIBONACCI")
    print("-" * 50)

    n = 35
    print(f"   Calculando F({n})...")

    inicio = time.time()
    r1 = fib_recursivo(n)
    t1 = time.time() - inicio

    inicio = time.time()
    r2 = fib_memo(n)
    t2 = time.time() - inicio

    print(f"   Recursivo: F({n}) = {r1} en {t1:.4f}s")
    print(f"   Memoizado: F({n}) = {r2} en {t2:.6f}s")
    if t2 > 0:
        print(f"   Speedup: {t1/t2:.0f}x más rápido")

    # ── Ejemplo 2: Escaleras ──
    print("\n2. SUBIR ESCALERAS")
    print("-" * 50)
    for n in [5, 10, 20, 30]:
        resultado = escaleras_memo(n)
        print(f"   n={n:2d} → {resultado:,} formas")

    # ── Ejemplo 3: Monedas ──
    print("\n3. CAMBIO DE MONEDAS")
    print("-" * 50)
    casos = [
        ([1, 5, 10], 12),
        ([1, 3, 4], 6),
        ([1, 5, 10, 25], 63),
    ]
    for monedas, monto in casos:
        resultado = monedas_memo(monedas, monto)
        print(f"   Monedas {monedas}, monto={monto} → {resultado} monedas")

    # Mostrar que greedy falla
    print("\n   ⚠ Greedy vs DP en monedas [1, 3, 4], monto=6:")
    print("     Greedy: 4+1+1 = 3 monedas (toma la más grande primero)")
    print("     DP:     3+3   = 2 monedas (solución óptima)")

    # ── Ejemplo 4: Mochila ──
    print("\n4. MOCHILA 0/1")
    print("-" * 50)
    pesos = [2, 3, 4, 5]
    valores = [3, 4, 5, 6]
    capacidad = 8

    print(f"   Objetos: peso={pesos}, valor={valores}")
    print(f"   Capacidad: {capacidad}")

    inicio = time.time()
    r1 = mochila_recursivo(pesos, valores, capacidad)
    t1 = time.time() - inicio

    inicio = time.time()
    r2 = mochila_memo(pesos, valores, capacidad)
    t2 = time.time() - inicio

    print(f"   Recursivo: valor={r1} en {t1:.6f}s")
    print(f"   Memoizado: valor={r2} en {t2:.6f}s")

    # ── Ejemplo 5: LCS ──
    print("\n5. SUBSECUENCIA COMÚN MÁS LARGA (LCS)")
    print("-" * 50)
    casos = [
        ("ABCBDAB", "BDCAB"),
        ("GATO", "GRITO"),
        ("ALGORITMO", "ALTRUISMO"),
    ]
    for s1, s2 in casos:
        resultado = lcs_memo(s1, s2)
        print(f'   "{s1}" vs "{s2}" → LCS = {resultado}')

    # ── Resumen ──
    print("\n" + "=" * 70)
    print("RESUMEN")
    print("=" * 70)
    print("""
    ┌─────────────────────┬──────────────┬──────────────┬──────────────┐
    │ Problema            │ Sin DP       │ Con Memo     │ Clave memo   │
    ├─────────────────────┼──────────────┼──────────────┼──────────────┤
    │ Fibonacci           │ O(2^n)       │ O(n)         │ n            │
    │ Escaleras           │ O(2^n)       │ O(n)         │ n            │
    │ Cambio de monedas   │ O(k^n)       │ O(n × k)     │ monto        │
    │ Mochila 0/1         │ O(2^n)       │ O(n × W)     │ (i, cap)     │
    │ LCS                 │ O(2^(n+m))   │ O(n × m)     │ (i, j)       │
    └─────────────────────┴──────────────┴──────────────┴──────────────┘

    Patrón de memoization:
    1. Escribir la solución recursiva normal
    2. Agregar un diccionario memo como parámetro
    3. Antes de calcular: if clave in memo → return memo[clave]
    4. Después de calcular: memo[clave] = resultado
    """)
```
