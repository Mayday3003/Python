## 🧮 72. Edit Distance

Las operaciones permitidas son:

* ✏️ **Insertar** un carácter
* ❌ **Eliminar** un carácter
* 🔁 **Reemplazar** un carácter

---

### 🧩 **Entrada**

`word1`, `word2`

### 📤 **Salida**

Número mínimo de operaciones para convertir `word1` en `word2`.

---

### 🧠 **Ejemplo 1**

```python
Input:
word1 = "horse"
word2 = "ros"

Output:
3
```

**Explicación:**
`horse → rorse` (reemplazar 'h' por 'r')
`rorse → rose` (eliminar 'r')
`rose → ros` (eliminar 'e')
→ Total = **3 operaciones**

---

### 🧠 **Ejemplo 2**

```python
Input:
word1 = "intention"
word2 = "execution"

Output:
5
```

**Explicación:**
`intention → inention` (eliminar 't')
`inention → enention` (reemplazar 'i' por 'e')
`enention → exention` (reemplazar 'n' por 'x')
`exention → exection` (reemplazar 'n' por 'c')
`exection → execution` (insertar 'u')
→ Total = **5 operaciones**

## 🥚🏢 Problema: Determinar el Piso Crítico con Huevos

Tienes `k` huevos **idénticos** y un edificio de `n` pisos, numerados del `1` al `n`.

Se sabe que existe un piso crítico `f` tal que:

* Si lanzas un huevo desde un piso **mayor** que `f`, el huevo **se rompe**.
* Si lo lanzas desde el piso `f` o desde un piso **inferior**, el huevo **no se rompe**.

Tu objetivo es **determinar con certeza** cuál es ese piso `f`.

### ✅ Reglas

En cada movimiento puedes:

* Tomar un huevo **sin romper** y lanzarlo desde cualquier piso `x` (1 ≤ x ≤ n)
* Si el huevo **se rompe**, lo pierdes y no puedes volver a usarlo.
* Si **no se rompe**, puedes reutilizarlo más adelante.

Debes encontrar el **mínimo número de movimientos necesarios** para garantizar que puedes encontrar `f`, sin importar cuál sea su valor.

### 🎯 Salida

Devuelve el **mínimo número de lanzamientos** necesarios para estar 100% seguro del valor de `f`.

---

### 📘 Ejemplos

#### Ejemplo 1

**Entrada:** `k = 1`, `n = 2`
**Salida:** `2`

**Explicación:**

* Intentas con el piso 1:

  * Si se rompe → `f = 0`
  * Si no se rompe, intentas piso 2:

    * Si se rompe → `f = 1`
    * Si no → `f = 2`

Se requieren mínimo **2 intentos**.

---

#### Ejemplo 2

**Entrada:** `k = 2`, `n = 6`
**Salida:** `3`

---

#### Ejemplo 3

**Entrada:** `k = 3`, `n = 14`
**Salida:** `4`


---


# Guía DP (bottom-up)

## 1) Idea clave (formulación por “movimientos”)

En lugar de preguntar: *“¿cuántos lanzamientos necesito para `n` pisos y `k` huevos?”*,
planteamos: **“con `k` huevos y `m` movimientos, ¿cuántos pisos puedo cubrir con certeza?”**
Si con `m` movimientos puedo cubrir ≥ `n` pisos, entonces `m` es suficiente.

## 2) Definición de estado

Sea `dp[m][k]` = **máximo número de pisos** que puedo determinar con certeza (cubrir) usando **k** huevos en **m** movimientos.

## 3) Transición (recurrencia)

Al lanzar desde un piso “pivote”:

* Si **se rompe**, me quedan `k-1` huevos y `m-1` movimientos → cubro `dp[m-1][k-1]` pisos **debajo**.
* Si **no se rompe**, me quedan `k` huevos y `m-1` movimientos → cubro `dp[m-1][k]` pisos **arriba**.
  Sumo además el piso de prueba actual: `+ 1`.

**Recurrencia:**

```
dp[m][k] = dp[m-1][k-1] + dp[m-1][k] + 1
```

## 4) Condiciones iniciales

* `dp[0][k] = 0` para todo `k`  (con 0 movimientos, cubro 0 pisos).
* `dp[m][0] = 0` para todo `m`  (con 0 huevos, no puedo cubrir nada).

## 5) Construcción bottom-up (procedimiento)

1. Inicializa una tabla `dp` de tamaño `(M+1) x (k+1)` con ceros (no sabemos aún `M`, lo iremos incrementando).
2. Para `m = 1, 2, 3, ...`:

   * Para `eggs = 1..k`:

     * Calcula `dp[m][eggs] = dp[m-1][eggs-1] + dp[m-1][eggs] + 1`.
   * Si en cualquier punto `dp[m][k] >= n`, **termina**: ese `m` es la respuesta mínima.

> Intuición: `dp[m][k]` crece muy rápido; por eso basta aumentar `m` hasta cubrir `n`.

## 6) Correctitud (invariante)

Para cada `(m, k)`, `dp[m][k]` cuenta exactamente la **cobertura garantizada** sumando:

* lo que cubro si rompe (abajo) + el piso de prueba + lo que cubro si no rompe (arriba).
  La construcción bottom-up garantiza que al evaluar `(m, k)` ya conozco `(m-1, k-1)` y `(m-1, k)`.

## 7) Complejidad

* Tiempo: **O(k · M)**, donde `M` es la respuesta (número mínimo de movimientos).
  En la práctica es muy eficiente (para `n` grande, `M` es ≈ `O(log n)`–`O(√n)` según `k`).
* Espacio: **O(k)** si optimizas por filas; **O(k·M)** si guardas toda la tabla.

## 8) Mini-tabla de ejemplo (k=2, n=6)

* `m=1`: dp[1][1]=1, dp[1][2]=2
* `m=2`: dp[2][1]=2, dp[2][2]=4
* `m=3`: dp[3][1]=3, dp[3][2]=7  → como `dp[3][2]=7 ≥ 6`, respuesta = **3**.

---

# (Opcional) DP clásica por pisos (menos eficiente)

Otra formulación define `F[k][n]` = mín. movimientos para `k` huevos y `n` pisos, probando un piso `x`:

```
F[k][n] = 1 + min_x( max( F[k-1][x-1], F[k][n-x] ) )
```

Es bottom-up pero **O(k·n·log n)** (con bsearch) o **O(k·n²)**; buena para enseñar, menos práctica para `n` grande.

> Tip didáctico: Pide a los estudiantes construir la tabla `dp[m][k]` por filas y detenerse cuando `dp[m][k] ≥ n`. Es muy visual y conecta con “árbol de decisiones” → “tabla DP”.

````python
# Versión discutida en clase

from tabulate import tabulate as t
def super_egg(k: int, n: int):
  if n == 0:
    return [[0]*(n+1) for _ in range(k+1)], 0
  if k == 0:
    # problema inválido (sin huevos no se puede determinar nada); devolvemos 0 movs
    return [[0]*(n+1) for _ in range(k+1)], 0

  # dp[e][f] = min lanzamientos con e huevos y f pisos
  dp = [[0]*(n+1) for _ in range(k+1)]

  # bases:
  #  - 0 pisos => 0 lanzamientos
  #  - 1 piso  => 1 lanzamiento
  for e in range(1, k+1):
    dp[e][1] = 1
    dp[e][0] = 0

  # 1 huevo => f lanzamientos (búsqueda lineal)
  for f in range(1, n+1):
    dp[1][f] = f

  # transición O(k·n²)
  for e in range(2, k+1):
    for f in range(2, n+1):
      best = float('inf')
      # probar cada piso x como primer lanzamiento
      for x in range(1, f+1):
        worst = 1 + max(dp[e-1][x-1], dp[e][f-x])  # rompe vs no rompe
        if worst < best:
          best = worst
      dp[e][f] = best

  return dp, dp[k][n]


#print(t(super_egg(k=1, n=1)[0])) # 1
#print(t(super_egg(k=1, n=0)[0])) # 0
#print(t(super_egg(k=1, n=1)[0])) # 1
#print(t(super_egg(k=1, n=2)[0])) # 2
#print(t(super_egg(k=1, n=10)[0])) # 10
#print(t(super_egg(k=2, n=1)[0])) # 1
#print(t(super_egg(k=2, n=2)[0])) # 2
#print(t(super_egg(k=2, n=6)[0])) # 3
#print(t(super_egg(k=2, n=7)[0])) # 4
#print(t(super_egg(k=3, n=14)[0])) # 4
#print(t(super_egg(k=3, n=25)[0])) # 5
#print(t(super_egg(k=3, n=100)[0])) # 9
#print(t(super_egg(k=4, n=100)[0])) # 8
print(t(super_egg(k=10, n=100)[0])) # 7
#print(t(super_egg(k=2, n=100)[0])) # 14
#print(t(super_egg(k=5, n=500)[0])) # 10
````
