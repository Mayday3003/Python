## 1. Descripción General

El trabajo final consiste en seleccionar un **problema del mundo real** y resolverlo aplicando progresivamente cada una de las técnicas vistas en el curso. El objetivo es que el estudiante demuestre no solo el dominio técnico de cada paradigma algorítmico, sino también la capacidad de **comparar, analizar y justificar** cuándo conviene usar cada enfoque.

## 3. Casos de Uso Sugeridos

Cada equipo debe elegir **uno** de los siguientes casos de uso, o proponer uno propio (sujeto a aprobación del profesor). El problema elegido debe ser lo suficientemente completo para aplicar las 5 técnicas del curso.

### 3.4 Optimización de Tareas en Servidores

Un sistema de cómputo tiene `m` servidores y recibe `n` tareas con tiempos de ejecución, prioridades y dependencias. Se busca minimizar el tiempo total de ejecución.

**Subproblemas:**
- Asignar tareas a servidores (scheduling)
- Respetar dependencias entre tareas
- Balancear carga entre servidores

## 4. Estructura de Entregas

### Entrega 1 — Análisis de Complejidad y Recursividad


**Contenido:**

1. **Definición del problema**
   - Descripción clara del caso de uso elegido
   - Identificación de las entradas, salidas y restricciones
   - Conjunto de datos de prueba (mínimo 3 casos: pequeño, mediano, grande)

2. **Solución por fuerza bruta**
   - Implementación funcional que resuelva el problema sin optimizaciones
   - Código documentado y ejecutable

3. **Análisis de complejidad**
   - Complejidad temporal y espacial de la solución por fuerza bruta (Big O)
   - Justificación paso a paso del análisis
   - Medición empírica de tiempos con diferentes tamaños de entrada
   - Gráfica comparativa (tiempo vs tamaño de entrada)

4. **Solución recursiva**
   - Reimplementación usando recursividad
   - Identificación clara del caso base y caso recursivo
   - Análisis de complejidad de la versión recursiva
   - Comparación de tiempos contra la versión iterativa/fuerza bruta

**Entregables:**
- Código fuente (`.py`)
- Documento con análisis y gráficas (`.pdf`)

---

### Entrega 2 — Greedy y Backtracking


**Contenido:**

1. **Solución Greedy**
   - Identificar la "decisión local óptima" para el problema
   - Implementación del algoritmo greedy
   - Análisis de complejidad
   - Análisis crítico: ¿la solución greedy es óptima para este problema? ¿En qué casos falla? Incluir al menos un contraejemplo si no es óptima

2. **Solución con Backtracking**
   - Implementación que explore el espacio de soluciones con poda
   - Identificar qué restricciones permiten podar ramas
   - Análisis de complejidad (peor caso vs caso promedio con poda)
   - Comparación de tiempos contra las soluciones anteriores

3. **Comparativa parcial**
   - Tabla comparativa de las 4 soluciones hasta el momento (fuerza bruta, recursiva, greedy, backtracking)
   - Discusión de trade-offs: ¿cuál es más rápida? ¿cuál da mejor resultado? ¿cuál es más fácil de implementar?

**Entregables:**
- Código fuente (`.py`)
- Documento con análisis y comparativas (`.pdf`)

---

### Entrega 3 — Divide y Vencerás + Comparativa Final


**Contenido:**

1. **Solución con Divide y Vencerás**
   - Identificar cómo dividir el problema en subproblemas independientes
   - Implementación del algoritmo
   - Explicar claramente las fases: dividir, vencer, combinar
   - Análisis de complejidad (usar Teorema Maestro si aplica)

2. **Comparativa final**
   - Tabla comparativa de las 5 técnicas aplicadas al problema:

   |---|---|---|---|---|---|
   | Fuerza bruta | | | | | |
   | Recursividad | | | | | |
   | Greedy | | | | | |
   | Backtracking | | | | | |
   | Divide y Vencerás | | | | | |

   - Gráfica de tiempos de ejecución de las 5 soluciones
   - Análisis: ¿qué técnica es la más adecuada para este problema y por qué?

3. **Conclusiones**
   - Reflexión sobre qué aprendieron al aplicar cada técnica al mismo problema
   - ¿En qué escenarios reales usarían cada paradigma?
   - Limitaciones encontradas

**Entregables:**
- Código fuente completo y organizado (`.py`)
- Documento final con todo el análisis (`.pdf`)
