# ðŸ§  ConceptualizaciÃ³n: Divide and Conquer (Dividir y Conquistar)



## ðŸ“Œ Â¿QuÃ© es?

**Divide and Conquer** es un **paradigma de diseÃ±o de algoritmos** que consiste en resolver un problema dividiÃ©ndolo en subproblemas mÃ¡s pequeÃ±os, resolviendo cada uno de forma independiente (posiblemente de manera recursiva) y combinando las soluciones para obtener la respuesta al problema original.

---

## ðŸ§© Etapas del paradigma

Todo algoritmo basado en Divide and Conquer sigue tres fases fundamentales:

1. **Divide ðŸª“**
   âž¤ Separar el problema original en dos o mÃ¡s subproblemas mÃ¡s pequeÃ±os del mismo tipo.

2. **Conquer âš”ï¸**
   âž¤ Resolver recursivamente los subproblemas.

3. **Combine ðŸ§µ**
   âž¤ Integrar las soluciones de los subproblemas para obtener la soluciÃ³n del problema original.

## ðŸ” Â¿Por quÃ© se usa recursiÃ³n?

La recursiÃ³n se adapta naturalmente a este paradigma, ya que permite **aplicar la misma estrategia a subproblemas mÃ¡s pequeÃ±os**, haciendo que el algoritmo se auto-replique hasta llegar al caso base.

| ------------------------- | ------------------------------------ | ------------------------- | ----------------------------- |
| ðŸ§® Merge Sort             | Dividir arreglo en mitades           | Ordenar cada mitad        | Mezclar (merge) ordenadamente |
| ðŸ” Binary Search          | Mitad izquierda o derecha            | Buscar en mitad relevante | Retornar el resultado         |
| ðŸ“Š Quick Sort             | Elegir pivote y dividir              | Ordenar subconjuntos      | Combinar (implÃ­citamente)     |

## ðŸ“ˆ AnÃ¡lisis de complejidad

## ðŸŽ¯ Â¿CuÃ¡ndo es Ãºtil?

## âš ï¸ Limitaciones

## ðŸ§¾ DescripciÃ³n general de Merge Sort

**Merge Sort** es un algoritmo de ordenamiento eficiente, **estable** y con una complejidad garantizada de $\mathcal{O}(n \log n)$ en el peor caso.

````python

def merge(data: list[int], left: int, right: int, mid: int) -> None:
  n_left = mid-left+1
  n_right = right-mid
  left_copy = [data[left+i] for i in range(n_left)]
  right_copy = [data[mid+i+1] for i in range(n_right)]

  i = 0
  j = 0
  k = left
  while(i < n_left and j < n_right):
    if(left_copy[i] <= right_copy[j]):
      data[k] = left_copy[i]
      i += 1
    else:
      data[k] = right_copy[j]
      j += 1
    k += 1

  while(i < n_left):
    data[k] = left_copy[i]
    i += 1
    k += 1

  while(j < n_right):
    data[k] = right_copy[j]
    j += 1
    k += 1

def mergesort(data: list[int], left: int = 0, right: int = 0) -> None:
  if(left < right):
    mid = (left + right) // 2
    mergesort(data, left, mid)
    mergesort(data, mid+1, right)
    merge(data, left, right, mid)

data = [1,2,3,4,5,6,3,7,8,2,9,1,10]
print(id(data), data)
mergesort(data, 0, len(data)-1)
print(id(data), data)
````

## ðŸ§¾ Â¿QuÃ© es QuickSort?

**QuickSort** es un algoritmo de ordenamiento muy eficiente y elegante que:

* Usa el paradigma **Divide and Conquer**.
* Tiene una complejidad promedio de $\mathcal{O}(n \log n)$.
* En la prÃ¡ctica, suele ser mÃ¡s rÃ¡pido que Merge Sort por su **mejor uso de memoria** y **menores costos de combinaciÃ³n**.

> Quicksort es como un guerrero Ã¡gil: no guarda listas auxiliares, solo **divide rÃ¡pido, ataca eficazmente** y sigue adelante.

---

## ðŸŽ¯ Â¿CuÃ¡l es la idea principal?

1. **Elegir un elemento como pivote**.
2. **Dividir** el arreglo en dos grupos:

   * Los elementos **menores** al pivote.
   * Los elementos **mayores** al pivote.
3. **Ordenar recursivamente** cada grupo.
4. **Unir** los resultados (en algunas versiones, esto es implÃ­cito).

---

## ðŸ”Ž Â¿Por quÃ© es Divide and Conquer?

| Fase           | AcciÃ³n en QuickSort                                               |
| -------------- | ----------------------------------------------------------------- |
| **DIVIDE** ðŸª“  | Elegir un pivote y separar en elementos menores y mayores.        |
| **CONQUER** âš”ï¸ | Ordenar recursivamente los subarreglos.                           |
| **COMBINE** ðŸ§µ | Concatenar los resultados ordenados (menores + pivote + mayores). |

ðŸŽ¯ A diferencia de Merge Sort, **la combinaciÃ³n en QuickSort es simple** (sÃ³lo una concatenaciÃ³n).

---

## ðŸ’¡ Ejemplo paso a paso

Supongamos que queremos ordenar:

```python

```

### Paso 1: Elegir pivote

* Mayores: `[8, 10]`

### Paso 2: RecursiÃ³n

* Ordenamos `[3, 1, 0, 2]` (nuevo pivote: 2) â†’ Menores: `[1, 0]`, Mayores: `[3]`
* Ordenamos `[8, 10]` â†’ ya estÃ¡n en orden

### Paso 3: Combinar

* `[0, 1] + [2] + [3]` â†’ `[0, 1, 2, 3]`
* Luego: `[0, 1, 2, 3] + [7] + [8, 10]` â†’ `[0, 1, 2, 3, 7, 8, 10]`

---

## ðŸ§  Â¿Por quÃ© es tan eficiente?

* **No necesita estructuras auxiliares grandes** como Merge Sort.
* En el **mejor y promedio caso**, cada particiÃ³n divide el arreglo en partes similares â†’ $\mathcal{O}(n \log n)$.
* Pero âš ï¸ en el **peor caso** (pivote mal elegido), puede volverse $\mathcal{O}(n^2)$.

ðŸ’¡ Por eso, **la elecciÃ³n del pivote es crÃ­tica**. Estrategias comunes:

* Mediana de tres (mejor para evitar sesgos).

## ðŸ§© Â¿CuÃ¡ndo usar QuickSort?

âœ… Cuando necesitas **velocidad y eficiencia en memoria**.
