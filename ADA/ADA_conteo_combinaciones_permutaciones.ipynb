# 🧮 Conteo - Parte 2: Principios Fundamentales

## 🎯 Objetivo de esta sección

Conocer y aplicar los **dos principios básicos** que permiten contar de manera sistemática:

* ➕ El **principio aditivo**
* ✖️ El **principio multiplicativo**

---

## ✴️ 1. Principio Aditivo (o de la suma)

> **Cuando una tarea puede realizarse de varias formas *mutuamente excluyentes*, el total de formas posibles es la suma de las formas de cada caso.**

🔎 *Mutuamente excluyente* = no se pueden hacer al mismo tiempo.

### 🎓 Ejemplo:

> Puedes elegir entre ir al cine 🎥 o al teatro 🎭
>
> * Hay 4 funciones de cine
> * Hay 3 funciones de teatro

**¿Cuántas opciones distintas tienes?**

$$
4 + 3 = 7
$$

---

## ✴️ 2. Principio Multiplicativo (o de la multiplicación)

> **Cuando una tarea se realiza en etapas, y para cada elección hay un número fijo de formas de continuar, el total de formas posibles es el producto de las formas de cada etapa.**

🔧 Cada elección **depende de una anterior**, pero las elecciones no interfieren entre sí.

### 🎓 Ejemplo:

> Quieres elegir un conjunto de ropa.
>
> * Tienes 3 camisas 👕
> * Tienes 2 pantalones 👖

**¿Cuántos outfits distintos puedes formar?**

$$
3 \times 2 = 6
$$

---

## 🤝 Comparación rápida

| Situación                                            | ¿Suma o Multiplicación? | Ejemplo                          |
| ---------------------------------------------------- | ----------------------- | -------------------------------- |
| Eliges **una entre varias alternativas excluyentes** | ➕ Suma                  | Cine (4) o teatro (3): 4 + 3     |
| Tomas **decisiones consecutivas**                    | ✖️ Multiplicación       | Camisa (3) y pantalón (2): 3 × 2 |

# 🔀 Conteo - Parte 3: Permutaciones

## 🎯 Objetivo de esta sección

* Comprender cuándo se usan permutaciones.
* Aplicar las fórmulas de permutación **con y sin repetición**.
* Resolver problemas reales donde el orden es relevante.

---

## ❓ ¿Qué es una permutación?

> Una **permutación** es una **forma de ordenar elementos**, donde **el orden importa**.

### ✳️ Ejemplo básico:

> ¿Cuántas formas hay de ordenar 4 libros diferentes en una repisa?

$$
4! = 4 \times 3 \times 2 \times 1 = 24
$$

---

## 🧪 Permutaciones **sin repetición**

> Tienes $n$ elementos **distintos**, y quieres ordenarlos.

### 🔸 Fórmula general:

$$
P(n) = n!
$$

---

### 🧩 Ejemplo 1:

> ¿Cuántas formas diferentes puedes acomodar a 5 personas en una fila?

$$
P(5) = 5! = 120
$$

---

### 🧩 Ejemplo 2:

> ¿Cuántas formas diferentes hay de asignar los primeros 3 puestos en una carrera con 10 corredores?

Esto es una **permutación de 3 elementos tomados de 10**:

$$
P(10, 3) = 10 \times 9 \times 8 = 720
$$

### 📌 Fórmula general:

$$
P(n, r) = \frac{n!}{(n - r)!}
$$

---

## 🔁 Permutaciones **con repetición**

> Cuando hay elementos **repetidos**, no todos los ordenamientos se consideran diferentes.

### 🔸 Fórmula:

$$
\text{Permutaciones con repetición} = \frac{n!}{k_1! \cdot k_2! \cdot \dots \cdot k_r!}
$$

Donde:

* $n$ es el total de elementos
* $k_i$ es la cantidad de veces que se repite cada elemento

---

### 🧩 Ejemplo 3:

> ¿Cuántas formas distintas hay de ordenar las letras de la palabra **MAMÁ**?

Hay 4 letras:

* M (2 veces), A (2 veces)

$$
\frac{4!}{2! \cdot 2!} = \frac{24}{4} = 6
$$

---

### 🧩 Ejemplo 4:

> ¿Cuántas permutaciones distintas tiene la palabra **BANANA**?

Letras: 6

* A (3 veces), N (2 veces), B (1 vez)

$$
\frac{6!}{3! \cdot 2! \cdot 1!} = \frac{720}{6 \cdot 2} = 60
$$

---

## 🧠 ¿Cómo saber si usar permutaciones?

📌 Pregúntate:

1. ¿Estoy **ordenando elementos**?
2. ¿Todos los elementos son **distintos**?
3. ¿Hay elementos **repetidos**?
4. ¿Estoy seleccionando solo algunos (r) de un total (n)?

---

## 🧪 Ejercicios para resolver en clase

### 🧩 Ejercicio 1:

> ¿Cuántas formas diferentes puedes formar una fila con 7 estudiantes?

👉 Orden importa, sin repetición.

$$
7! = 5040
$$

---

### 🧩 Ejercicio 2:

> Hay 5 libros y se deben elegir 3 para ubicarlos en una vitrina, en orden.

$$
P(5, 3) = \frac{5!}{(5-3)!} = 60
$$

---

### 🧩 Ejercicio 3:

> ¿Cuántas permutaciones distintas tiene la palabra **COLOSO**?

Letras: 6

* C (1), O (3), L (1), S (1)

$$
\frac{6!}{3!} = \frac{720}{6} = 120
$$

# 🧮 Conteo - Parte 4: Combinaciones

## 🎯 Objetivo de esta sección

* Comprender qué es una combinación.
* Aprender a usar la fórmula de combinaciones **sin repetición**.
* Comparar combinaciones con permutaciones.
* Resolver problemas prácticos en los que **el orden no importa**.

---

## ❓ ¿Qué es una combinación?

> Una **combinación** es una selección de elementos **sin importar el orden**.

---

### 📌 Diferencia clave

| Concepto    | ¿Importa el orden? | Ejemplo típico                   |
| ----------- | ------------------ | -------------------------------- |
| Permutación | ✅ Sí               | Acomodar libros, asignar puestos |
| Combinación | ❌ No               | Armar grupos, elegir comités     |

---

## 🔢 Fórmula de combinaciones sin repetición

Si tienes $n$ elementos y eliges $r$ **sin importar el orden**:

$$
C(n, r) = \binom{n}{r} = \frac{n!}{r!(n - r)!}
$$

---

### 🧩 Ejemplo 1

> ¿Cuántos grupos de 3 personas se pueden formar con 8 estudiantes?

$$
C(8, 3) = \frac{8!}{3! \cdot 5!} = \frac{8 \times 7 \times 6}{3 \times 2 \times 1} = 56
$$

👉 Si solo me interesa **quiénes están en el grupo**, no su orden, es combinación.

---

### 🧩 Ejemplo 2

> En un concurso, debes elegir 5 cartas de una baraja de 52. ¿Cuántas manos distintas puedes recibir?

$$
C(52, 5) = \frac{52!}{5! \cdot 47!} = 2,598,960
$$

---

## 🧪 Ejercicio guiado (a resolver en clase)

> ¿Cuántas formas hay de elegir 4 estudiantes de un grupo de 10 para representar la clase?

$$
C(10, 4) = \frac{10!}{4! \cdot 6!} = \frac{10 \times 9 \times 8 \times 7}{4 \times 3 \times 2 \times 1} = 210
$$

---

## 🔁 Comparación con permutaciones

> ¿Qué cambia si **sí importa el orden** en el ejemplo anterior?

Entonces usamos permutaciones:

$$
P(10, 4) = \frac{10!}{6!} = 5040
$$

🔄 ¡Misma situación, distinto resultado por cambiar el criterio de orden!

---

## 💡 Cómo identificar que es una combinación

✅ Pregúntate:

* ¿Estoy **eligiendo un grupo** o **un subconjunto**?
* ¿No me importa el orden?
* ¿No repito elementos?

Si respondes “sí” → usa **combinaciones**.

### 🧠 Ejercicio 2. Combinaciones de r elementos entre n

> Dado un conjunto de elementos y un número $r$, imprime todas las combinaciones posibles **sin importar el orden**.

#### Entrada:

```python
['a', 'b', 'c', 'd'], r = 2
```

#### Salida esperada:

```
['a', 'b']
['a', 'c']
['a', 'd']
['b', 'c']
['b', 'd']
['c', 'd']
```

🔍 *A tener en cuenta*: Plantea una solución usando recursión y otra usando la función `combinations` de `itertools`.



````python
def combinaciones_r(lista: list[str], r: int = 0,
                    rta_actual: list[str] = [],
                    rtas: list[list[str]] = []) -> list[list[str]]:

  #caso base
  if(len(rta_actual) == r):
    rta_actual = sorted(rta_actual) #ordenar
    if(rta_actual not in rtas):
      rtas.append(rta_actual.copy())
    return rtas

  #caso recursivo
  for opcion in lista:
    if(opcion in rta_actual):
      continue
    rta_actual.append(opcion)
    combinaciones_r(lista, r, rta_actual, rtas)
    rta_actual.pop()

  return rtas

combinaciones_r([1,2,3], r=2)
````

# Pastel

---
## 🔢 Principios Fundamentales

### 🟩 **Principio Aditivo**

Si una tarea puede realizarse de *n* maneras **o** de *m* maneras (mutuamente excluyentes):

$$
\text{Total} = n + m
$$

---

### 🟦 **Principio Multiplicativo**

Si una tarea se realiza en *k* etapas, con $n_1, n_2, ..., n_k$ formas de realizar cada una:

$$
\text{Total} = n_1 \times n_2 \times \dots \times n_k
$$

---

## 🔄 Permutaciones

### ✅ **Permutaciones simples (sin repetición)**

Número de formas de ordenar *n* elementos distintos:

$$
P(n) = n!
$$

---

### 🔁 **Permutaciones de *r* elementos de *n* (sin repetición)**

$$
P(n, r) = \frac{n!}{(n - r)!}
$$

---

### 🔂 **Permutaciones con repetición**

Si hay repeticiones de ciertos elementos:

$$
\text{Permutaciones} = \frac{n!}{k_1! \cdot k_2! \cdot \dots \cdot k_r!}
$$

Donde:

* $n$: total de elementos
* $k_i$: número de veces que se repite cada elemento

---

## 🎯 Combinaciones

### 📌 **Combinaciones sin repetición**

Número de subconjuntos de tamaño *r* que se pueden formar de un conjunto de *n* elementos, **sin importar el orden**:

$$
C(n, r) = \binom{n}{r} = \frac{n!}{r!(n - r)!}
$$

---

### 🔁 **Combinaciones con repetición**

$$
C(n + r - 1, r) = \binom{n + r - 1}{r}
$$

Donde:

* $n$: tipos de elementos
* $r$: cantidad de elementos a elegir (permitiendo repeticiones)

---

## 🧠 Recordatorios clave

| Tipo de problema           | ¿Orden importa? | ¿Repetición? | Fórmula                       |
| -------------------------- | --------------- | ------------ | ----------------------------- |
| Permutación total          | ✅ Sí            | ❌ No         | $n!$                          |
| Permutación parcial        | ✅ Sí            | ❌ No         | $\frac{n!}{(n-r)!}$           |
| Permutación con repetición | ✅ Sí            | ✅ Sí         | $\frac{n!}{k_1! \cdot \dots}$ |
| Combinación simple         | ❌ No            | ❌ No         | $\frac{n!}{r!(n-r)!}$         |
| Combinación con repetición | ❌ No            | ✅ Sí         | $\binom{n + r - 1}{r}$        |

---
