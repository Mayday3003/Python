# ğŸ§  CÃ³mo Elegir ParÃ¡metros y CuÃ¡ndo Usarlos  

### ğŸ¯ **Â¿Por quÃ© usar parÃ¡metros en una funciÃ³n?**  
Los parÃ¡metros permiten que una funciÃ³n sea **flexible** y **reutilizable**. Sin parÃ¡metros, una funciÃ³n siempre harÃ­a lo mismo, sin adaptarse a diferentes situaciones.

### ğŸ“Œ **Â¿CuÃ¡ntos parÃ¡metros definir?**

ğŸ”¹ **Regla general:** Usa **solo los parÃ¡metros necesarios** para que la funciÃ³n haga su tarea correctamente.

# ğŸ”„ Â¿CuÃ¡ndo Usar `return`?  

ğŸ”¹ `return` **devuelve un valor** para que pueda usarse despuÃ©s.  
ğŸ”¹ Si una funciÃ³n **no tiene `return`**, solo ejecuta cÃ³digo pero **no entrega un resultado**.

ğŸ“Œ **Ejemplo sin `return` (muestra un resultado, pero no lo guarda)**  
```python
def sumar(a, b):
    print(a + b)

resultado = sumar(4, 6)  # Output: 10
print(resultado)  # Output: None (no devuelve nada)
```

ğŸ“Œ **Ejemplo con `return` (devuelve un valor que se puede usar despuÃ©s)**  
```python
def sumar(a, b):
    return a + b

resultado = sumar(4, 6)  
print(resultado)  # Output: 10
```
âœ… AquÃ­, la funciÃ³n **devuelve el valor** y podemos almacenarlo en una variable (`resultado`).

## ğŸ¤” **Â¿CuÃ¡ndo usar `return`?**  
Usa `return` cuando **necesites que la funciÃ³n devuelva un resultado para usarlo despuÃ©s**.  

âœ” **Usa `return` en funciones que calculen algo**  
âœ” **Usa `return` en funciones que devuelvan informaciÃ³n especÃ­fica**  
âŒ **No uses `return` si la funciÃ³n solo imprime algo y no se necesita el valor despuÃ©s**  

**Ejemplo prÃ¡ctico:**  
```python
def obtener_doble(numero):
    return numero * 2

doble = obtener_doble(7)  
print(doble)  # Output: 14
```

## ğŸš¨ **Errores Comunes en Funciones y CÃ³mo Solucionarlos**

### âŒ **Error 1: Olvidar el `return` en una funciÃ³n**
ğŸ“Œ **Causa:** La funciÃ³n ejecuta su cÃ³digo pero **no devuelve nada**.

```python
def suma(a, b):
    resultado = a + b  # âŒ No retorna el resultado

print(suma(2, 3))  # None
```

âœ… **SoluciÃ³n:** Agregar `return` para devolver el resultado.
```python
def suma(a, b):
    return a + b  # âœ… Correcto

print(suma(2, 3))  # 5
```

---

### âŒ **Error 2: Llamar la funciÃ³n sin parÃ©ntesis**
ğŸ“Œ **Causa:** Llamar la funciÃ³n sin `()` devuelve una referencia a la funciÃ³n, no ejecuta su cÃ³digo.

```python
def mensaje():
    return "Hola"

print(mensaje)  # âŒ Muestra <function mensaje at 0x...>
```

âœ… **SoluciÃ³n:** Usar parÃ©ntesis `()` para llamar la funciÃ³n.
```python
print(mensaje())  # âœ… Muestra "Hola"
```

---

### âŒ **Error 3: Pasar un nÃºmero incorrecto de argumentos**
ğŸ“Œ **Causa:** Se pasan menos o mÃ¡s argumentos de los esperados.

```python
def resta(a, b):
    return a - b

print(resta(5))  # âŒ TypeError: missing 1 required argument
```

âœ… **SoluciÃ³n:** Verificar la cantidad de argumentos o definir valores por defecto.

```python
def resta(a, b=0):  # âœ… Asigna 0 si no se pasa el segundo argumento
    return a - b

print(resta(5))  # 5
```

---

### âŒ **Error 4: Modificar variables mutables dentro de una funciÃ³n**
ğŸ“Œ **Causa:** Python **no crea una nueva lista**, sino que modifica la original.

```python
def agregar_elemento(lista, elemento):
    lista.append(elemento)  # âŒ Modifica la lista original

mi_lista = [1, 2, 3]
agregar_elemento(mi_lista, 4)
print(mi_lista)  # âŒ Se modifica fuera de la funciÃ³n -> [1, 2, 3, 4]
```

âœ… **SoluciÃ³n:** Crear una copia de la lista dentro de la funciÃ³n.

```python
def agregar_elemento(lista, elemento):
    nueva_lista = lista[:]  # âœ… Copia la lista
    nueva_lista.append(elemento)
    return nueva_lista

mi_lista = [1, 2, 3]
nueva_lista = agregar_elemento(mi_lista, 4)
print(mi_lista)   # [1, 2, 3] (No se modificÃ³)
print(nueva_lista)  # [1, 2, 3, 4]
```

---

### âŒ **Error 5: Usar variables globales sin `global`**
ğŸ“Œ **Causa:** Python trata las variables globales como locales dentro de una funciÃ³n.

```python
contador = 0

def incrementar():
    contador += 1  # âŒ UnboundLocalError: la variable no estÃ¡ definida en la funciÃ³n

incrementar()
```

âœ… **SoluciÃ³n 1:** Usar `global` para modificar la variable global.

```python
contador = 0

def incrementar():
    global contador  # âœ… Correcto
    contador += 1

incrementar()
print(contador)  # 1
```

âœ… **SoluciÃ³n 2:** Pasar la variable como argumento.

```python
def incrementar(contador):
    return contador + 1

contador = incrementar(contador)
print(contador)  # 1
```

---

### âŒ **Error 6: No manejar `None` en las funciones**
ğŸ“Œ **Causa:** Intentar realizar operaciones con `None`.

```python
def dividir(a, b):
    if b == 0:
        return None  # âŒ Retorna None en caso de error

resultado = dividir(10, 0)
print(resultado + 5)  # âŒ TypeError: NoneType + int
```

âœ… **SoluciÃ³n:** Validar antes de operar.

```python
def dividir(a, b):
    if b == 0:
        return "Error: No se puede dividir por cero"  # âœ… Mejor manejo del error
    return a / b

resultado = dividir(10, 0)
print(resultado)  # Error: No se puede dividir por cero
```

---

### âŒ **Error 7: Olvidar `*args` o `**kwargs` en funciones flexibles**
ğŸ“Œ **Causa:** No permitir un nÃºmero variable de argumentos.

```python
def suma(a, b):
    return a + b

print(suma(1, 2, 3))  # âŒ TypeError: Too many arguments
```

âœ… **SoluciÃ³n:** Usar `*args` para manejar cualquier cantidad de argumentos.

```python
def suma(*numeros):
    return sum(numeros)  # âœ… Funciona con cualquier cantidad de nÃºmeros

print(suma(1, 2, 3, 4))  # 10
```

âœ… **Para argumentos con nombre, usar `**kwargs`**.
```python
def saludo(**kwargs):
    return f"Hola {kwargs.get('nombre', 'Invitado')}"

print(saludo(nombre="Carlos"))  # Hola Carlos
```

âœ” **Maneja variables mutables con cuidado (usa copias si es necesario).**  
âœ” **Si modificas variables globales, usa `global` o pÃ¡salas como argumentos.**  
âœ” **Valida tipos de datos con `type()` o `isinstance()`.**  
âœ” **Maneja `None` para evitar `TypeError`.**
