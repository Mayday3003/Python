# üìå Concepto de Pila (Stack)

Una **pila** (en ingl√©s, *Stack*) es una estructura de datos lineal que sigue el principio **LIFO** (*Last In, First Out*), lo que significa que el √∫ltimo elemento en entrar es el primero en salir.

Podemos imaginar una pila como una colecci√≥n de elementos apilados uno encima del otro, similar a un conjunto de platos en una mesa. Para agregar un nuevo plato (elemento), lo colocamos en la parte superior, y para retirar uno, debemos tomar el que est√° en la cima antes que los dem√°s.

---

### ‚öôÔ∏è **Caracter√≠sticas de una Pila**
1. **Orden de acceso:** Sigue la regla **LIFO (Last In, First Out)**.
2. **Operaciones principales:**
   - `Push`: Agregar un elemento a la cima de la pila.
   - `Pop`: Retirar el elemento superior de la pila.
   - `Peek` (o `Top`): Consultar el elemento en la cima sin retirarlo.
   - `IsEmpty`: Verificar si la pila est√° vac√≠a.
3. **Acceso restringido:** Solo podemos agregar o eliminar elementos desde la cima.
4. **Aplicaci√≥n en memoria:** Se implementa com√∫nmente usando **listas enlazadas** o **arreglos**.

---

### üèó **Casos de Uso de una Pila**
Las pilas son fundamentales en inform√°tica y se utilizan en diversas aplicaciones, tales como:

1. **Manejo de llamadas a funciones (Stack de ejecuci√≥n)**
   - Cuando una funci√≥n se invoca, se almacena en la pila hasta que finaliza.
   - Ejemplo: En lenguajes como C y Python, se usa la **pila de llamadas** para administrar la ejecuci√≥n de funciones de manera ordenada.

2. **Deshacer y rehacer acciones (Undo/Redo)**
   - Los editores de texto y los navegadores utilizan pilas para controlar cambios y permitir deshacer (undo) o rehacer (redo).

3. **Evaluaci√≥n de expresiones matem√°ticas**
   - La pila se usa en la evaluaci√≥n de expresiones en notaci√≥n **postfija (Notaci√≥n Polaca Inversa)** y en la conversi√≥n de expresiones infijas a postfijas.

4. **Sistemas de navegaci√≥n web (Historial de p√°ginas)**
   - Cuando navegamos por internet, las p√°ginas visitadas se almacenan en una pila, permitiendo retroceder a la anterior con el bot√≥n "atr√°s".

5. **Recorrido en estructuras de datos**
   - Se usa en algoritmos como **DFS (Depth-First Search)** para explorar grafos y √°rboles.

### üÜö Comparaci√≥n con otras estructuras de datos

### üöÄ **Conclusi√≥n**
Las **pilas (stacks)** son una estructura de datos simple pero poderosa, utilizada en m√∫ltiples aplicaciones del desarrollo de software y sistemas inform√°ticos. ### üî• **Resumen de las operaciones**
- `push(elemento)`: Apila un nuevo elemento en la cima.
- `peek()`: Muestra el elemento en la cima sin retirarlo.
- `pop()`: Quita el elemento superior de la pila.
- `isEmpty()`: Indica si la pila est√° vac√≠a.

```python
from dataclasses import dataclass

class EmptyQueueError(Exception):
    pass

@dataclass
class Stack: 
    data: list = []

    def push(self, value):
        self.data.append(value)

    def pop(self):
        if self.is_empty():
            raise EmptyQueueError("Stack is empty")
        return self.data.pop()

    def is_empty(self):
        return len(self.data) == 0

    def __repr__(self):
        return f"Stack({self.data})"
    
    def __len__(self):
        return len(self.data)
    
    
    
```
