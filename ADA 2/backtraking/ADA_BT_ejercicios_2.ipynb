# Ejercicio 1


Se te da un entero `n`. Devuelve la cantidad de números `x` con **dígitos únicos**, donde se cumple que `0 <= x < 10^n`.

### Ejemplo 1:

**Entrada:** `n = 2`
**Salida:** `91`
**Explicación:** La respuesta corresponde al total de números en el rango `0 ≤ x < 100`, excluyendo aquellos que repiten dígitos como `11, 22, 33, 44, 55, 66, 77, 88, 99`.

### Ejemplo 2:

**Entrada:** `n = 0`
**Salida:** `1`

### Restricciones:

* `0 <= n <= 8`

# Ejercicio 2

Se te da un string `s` y un diccionario de palabras `wordDict`.
Debes insertar espacios en `s` para construir oraciones en las que **cada palabra sea válida** dentro del diccionario.

Devuelve **todas las oraciones posibles** en cualquier orden.

👉 Nota: Una misma palabra del diccionario puede reutilizarse múltiples veces en la segmentación.

---

### Ejemplo 1:

**Entrada:**
`s = "catsanddog"`, `wordDict = ["cat","cats","and","sand","dog"]`
**Salida:**
`["cats and dog","cat sand dog"]`

### Ejemplo 2:

**Entrada:**
`s = "pineapplepenapple"`, `wordDict = ["apple","pen","applepen","pine","pineapple"]`
**Salida:**
`["pine apple pen apple","pineapple pen apple","pine applepen apple"]`
**Explicación:** Se permite reutilizar palabras del diccionario.

### Ejemplo 3:

**Entrada:**
`s = "catsandog"`, `wordDict = ["cats","dog","sand","and","cat"]`
**Salida:**
`[]`

---

### Restricciones:

* `1 <= s.length <= 20`
* `1 <= wordDict.length <= 1000`
* `1 <= wordDict[i].length <= 10`
* `s` y `wordDict[i]` consisten únicamente en letras minúsculas del inglés.
* Todas las palabras en `wordDict` son únicas.
* La entrada está construida de manera que la longitud de la respuesta no supere `10^5`.

# Ejercicio 4


Se te da una ecuación representada por una lista de palabras en el lado izquierdo y un resultado en el lado derecho.
Debes determinar si la ecuación es **resoluble** bajo las siguientes reglas:

1. Cada carácter se decodifica como un dígito del `0` al `9`.
2. No se permite que dos caracteres diferentes correspondan al mismo dígito.
3. Cada palabra `words[i]` y el `result` se interpretan como un número **sin ceros a la izquierda**.
4. La suma de los números obtenidos de `words` debe ser igual al número representado por `result`.

Devuelve `true` si la ecuación es resoluble, de lo contrario devuelve `false`.

---

### Ejemplo 1:

**Entrada:**
`words = ["SEND","MORE"], result = "MONEY"`
**Salida:**
`true`
**Explicación:** Asignación válida:
`'S' -> 9, 'E' -> 5, 'N' -> 6, 'D' -> 7, 'M' -> 1, 'O' -> 0, 'R' -> 8, 'Y' -> 2`
Con lo cual: `"SEND" + "MORE" = "MONEY"` → `9567 + 1085 = 10652`.

---

### Ejemplo 2:

**Entrada:**
`words = ["SIX","SEVEN","SEVEN"], result = "TWENTY"`
**Salida:**
`true`
**Explicación:** Una asignación posible es:
`'S' -> 6, 'I' -> 5, 'X' -> 0, 'E' -> 8, 'V' -> 7, 'N' -> 2, 'T' -> 1, 'W' -> 3, 'Y' -> 4`
De modo que: `"SIX" + "SEVEN" + "SEVEN" = "TWENTY"` → `650 + 68782 + 68782 = 138214`.

---

### Ejemplo 3:

**Entrada:**
`words = ["LEET","CODE"], result = "POINT"`
**Salida:**
`false`
**Explicación:** No existe una asignación de dígitos que cumpla con la ecuación.

---

### Restricciones:

* `2 <= words.length <= 5`
* `1 <= words[i].length, result.length <= 7`
* `words[i]` y `result` contienen solo letras mayúsculas del inglés.
* El número de caracteres distintos usados en la expresión es como máximo `10`.
