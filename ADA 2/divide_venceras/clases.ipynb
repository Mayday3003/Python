```python
def suma_divide_venceras(arr, inicio, fin):
    # Caso base: un solo elemento
    if inicio == fin:
        return arr[inicio]

    # Divide
    medio = (inicio + fin) // 2

    # Vence (resuelve subproblemas) y combina
    suma_izq = suma_divide_venceras(arr, inicio, medio)
    suma_der = suma_divide_venceras(arr, medio + 1, fin)

    return suma_izq + suma_der


def promedio_divide_venceras(arr):
    suma_total = suma_divide_venceras(arr, 0, len(arr) - 1)
    return suma_total / len(arr)


# Ejemplo de uso
numeros = [10, 20, 30, 40, 50]
print("Promedio:", promedio_divide_venceras(numeros))  # 30.0
```

```python
# ═══════════════════════════════════════════════════════════════════════════
# EJERCICIO 2: Encontrar el Máximo en un Arreglo Rotado
# ═══════════════════════════════════════════════════════════════════════════
#
# Problema: Un arreglo ordenado fue rotado k posiciones. Encontrar
#           el elemento máximo.
#
#           Un arreglo rotado es uno que estaba ordenado y se le
#           movieron los primeros k elementos al final.
#           Ej: [1,2,3,4,5] rotado 3 → [4,5,1,2,3]
#
# Ejemplo:
#   [4, 5, 6, 7, 1, 2, 3]  → 7
#   [3, 1, 2]               → 3
#   [1, 2, 3, 4, 5]         → 5  (sin rotación)
#   [2, 1]                  → 2
#
# Idea D&V:
#   - Si el subarreglo está ordenado (arr[izq] <= arr[der]),
#     el máximo es arr[der]
#   - Si no, el máximo está en la mitad que contiene el "quiebre"
#   - Comparar arr[medio] con arr[izq] para decidir qué mitad explorar
#
# Fuerza bruta: O(n)
# D&V: O(log n)
#

def maximo_rotado(arr: list[int], izq: int = 0, der = None):
    """Encuentra el máximo en un arreglo ordenado rotado."""

    if der is None: 
        der = len(arr) - 1

    if len(arr) == 1:
        return arr[0]
    
    if arr[izq] <= arr[der]:
        return arr[der]
    
    mid = (izq + der) // 2
    if arr[izq] > arr[mid]:
        return maximo_rotado(arr, izq, mid)
    else:
        maximo_actual = mid

        maximo_derecha = maximo_rotado(arr, mid + 1, der)

        if maximo_derecha > maximo_actual:
            return maximo_derecha
        else: 
            return maximo_actual




maximo_rotado([3, 1, 2])
```
