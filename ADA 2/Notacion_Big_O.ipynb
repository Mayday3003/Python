```python
import time
import matplotlib.pyplot as plt
from math import log2
```

```python
def constant_time(n):
    """O(1) - Tiempo constante"""
    return 1
```

```python
def logarithmic_time(n):
    """O(log n) - Tiempo logarítmico"""
    count = 0
    while n > 1:
        n //= 2
        count += 1
    return count
```

```python
def linear_time(n):
    """O(n) - Tiempo lineal"""
    count = 0
    for _ in range(n):
        count += 1
    return count
```

```python
def linearithmic_time(n):
    """O(n log n) - Tiempo lineal-logarítmico"""
    count = 0
    for i in range(n):
        current = i + 1
        while current > 0:
            current //= 2
            count += 1
    return count
```

```python
def quadratic_time(n):
    """O(n^2) - Tiempo cuadrático"""
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count
```

```python
def measure_time(func, n):
    start = time.time()
    func(n)
    end = time.time()
    return end - start
```

```python
def plot_complexities(n_values):
    complexities = {
        'O(1)': constant_time,
        'O(log n)': logarithmic_time,
        'O(n)': linear_time,
        'O(n log n)': linearithmic_time,
        'O(n^2)': quadratic_time
    }

    plt.figure(figsize=(12, 8))
    for name, func in complexities.items():
        times = [measure_time(func, n) for n in n_values]
        plt.plot(n_values, times, label=name)

    plt.xlabel('Tamaño de entrada (n)')
    plt.ylabel('Tiempo de ejecución (segundos)')
    plt.title('Comparación de Complejidades Algorítmicas')
    plt.legend()
    plt.xscale('log')
    plt.yscale('log')
    plt.grid(True)
    plt.show()
```

```python
# Evaluar y graficar las complejidades
n_values = [10, 100, 1000, 10000]
plot_complexities(n_values)

# Imprimir ejemplos de valores
print("Ejemplos de valores para n =", n_values[-1])
print(f"O(1): {constant_time(n_values[-1])}")
print(f"O(log n): {logarithmic_time(n_values[-1])}")
print(f"O(n): {linear_time(n_values[-1])}")
print(f"O(n log n): {linearithmic_time(n_values[-1])}")
print(f"O(n^2): {quadratic_time(n_values[-1])}")
```

```python
def count_even(arr):
    count = 0                      # O(1) - Inicialización
    for num in arr:                # O(n) - Recorre cada elemento una vez
        if num % 2 == 0:           # O(1) - Operación de módulo, tiempo constante
            count += 1             # O(1) - Incremento simple
    return count                   # O(1) - Retorno final

# Complejidad total: O(n)
# Se calcula tomando la complejidad dominante, que es el bucle O(n)
```

```python
def linear_search(arr, target):
    for i in range(len(arr)):  # O(n) - Este bucle se ejecuta n veces
        if arr[i] == target:   # O(1) - Comparación simple, tiempo constante
            return i           # O(1) - Retorno inmediato si se encuentra
    return -1                  # O(1) - Retorno si no se encuentra

# Complejidad total: O(n)
# Se calcula tomando la complejidad dominante, que es el bucle O(n)
```

```python
def find_max_min(arr):
    if not arr:                    # O(1) - Comprobación inicial
        return None, None
    max_val = min_val = arr[0]     # O(1) - Inicialización
    for num in arr[1:]:            # O(n-1) ≈ O(n) - Recorre n-1 elementos
        if num > max_val:          # O(1) - Comparación simple
            max_val = num          # O(1) - Asignación simple
        elif num < min_val:        # O(1) - Comparación simple
            min_val = num          # O(1) - Asignación simple
    return max_val, min_val        # O(1) - Retorno final

# Complejidad total: O(n)
# Se calcula considerando el bucle principal que domina la complejidad
```

```python
def sum_2d_matrix(matrix):
    total = 0                          # O(1) - Asignación inicial
    for i in range(len(matrix)):       # O(n) - Donde n es el número de filas
        for j in range(len(matrix[i])): # O(m) - Donde m es el número de columnas
            total += matrix[i][j]      # O(1) - Suma simple, tiempo constante

    return total                       # O(1) - Retorno final

# Complejidad total: O(n * m)
# Se calcula multiplicando las complejidades de los bucles anidados
```

```python
# Exponenciación rápida
def fast_power(base, exponent):
    # O(1): Comparación simple
    if exponent == 0:
        # O(1): Retorno simple
        return 1

    # O(log n): Llamada recursiva, se ejecuta log n veces
    half = fast_power(base, exponent // 2)

    # O(1): Operación módulo y comparación
    if exponent % 2 == 0:
        # O(1): Multiplicación y retorno
        return half * half
    # O(1): Este else es implícito
    else:
        # O(1): Multiplicaciones y retorno
        return half * half * base

# Complejidad total: O(log n)
# La complejidad está dominada por las llamadas recursivas,
# que ocurren log n veces, cada una con operaciones O(1).
```

```python
# Encontrar el bit más significativo
def most_significant_bit(n):
    # O(1): Comparación simple
    if n == 0:
        # O(1): Retorno simple
        return 0

    # O(1): Asignación simple
    msb = 0
    # O(log n): El bucle se ejecuta log n veces
    while n > 1:
        # O(1): Operación de desplazamiento de bits
        n = n >> 1
        # O(1): Incremento simple
        msb += 1

    # O(1): Retorno simple
    return msb

# Complejidad total: O(log n)
# La complejidad está dominada por el bucle while, que se ejecuta log n veces,
# y cada iteración realiza operaciones O(1).
```

```python
def most_frequent(arr):
    count_dict = {}                    # O(1) - Creación del diccionario
    for num in arr:                    # O(n) - Recorre cada elemento una vez
        if num in count_dict:          # O(1) - Búsqueda en diccionario (promedio)
            count_dict[num] += 1       # O(1) - Incremento simple
        else:
            count_dict[num] = 1        # O(1) - Asignación simple

    max_count = 0                      # O(1) - Inicialización
    most_freq = None                   # O(1) - Inicialización
    for num, count in count_dict.items(): # O(k) - k es el número de elementos únicos
        if count > max_count:          # O(1) - Comparación simple
            max_count = count          # O(1) - Asignación simple
            most_freq = num            # O(1) - Asignación simple

    return most_freq                   # O(1) - Retorno final

# Complejidad total: O(n + k), donde n es el tamaño del array y k el número de elementos únicos
# Se calcula sumando la complejidad del primer bucle O(n) y el segundo O(k)
```

```python
def fibonacci(n):
    if n <= 1:                 # O(1) - Verificación de caso base
        return n               # O(1) - Retorno inmediato para casos base
    return fibonacci(n-1) + fibonacci(n-2)  # O(2^n) - Dos llamadas recursivas por cada n

# Complejidad total: O(2^n)
# Se calcula observando que cada llamada genera dos nuevas llamadas,
# formando un árbol binario de profundidad n
```

```python
def binary_search(arr, target):
    left = 0                           # O(1) - Inicialización
    right = len(arr) - 1               # O(1) - Inicialización

    while left <= right:               # O(log n) - El espacio de búsqueda se reduce a la mitad en cada iteración
        mid = (left + right) // 2      # O(1) - Cálculo simple
        if arr[mid] == target:         # O(1) - Comparación simple
            return mid                 # O(1) - Retorno si se encuentra
        elif arr[mid] < target:        # O(1) - Comparación simple
            left = mid + 1             # O(1) - Ajuste simple
        else:
            right = mid - 1            # O(1) - Ajuste simple

    return -1                          # O(1) - Retorno si no se encuentra

# Complejidad total: O(log n)
# Se calcula observando que en cada iteración, el espacio de búsqueda se reduce a la mitad
```

Algoritmo de Dijkstra (versión simplificada):

Este algoritmo encuentra el camino más corto desde un nodo inicial a todos los demás nodos en un grafo ponderado.
Puntos clave:

Utiliza una cola de prioridad (heap) para seleccionar eficientemente el siguiente nodo a procesar.
Mantiene un diccionario de distancias mínimas conocidas a cada nodo.

Análisis de complejidad:

Inicialización: O(V), donde V es el número de vértices.
Bucle principal: Se ejecuta potencialmente para cada arista, O(E) veces.
Operaciones de heap (heappush y heappop): O(log V) cada una.
Actualización de distancias: O(1) por cada vecino.

Complejidad total: O((V + E) log V)

En el peor caso, cada arista puede causar una actualización y una operación de heap.
Esta implementación es eficiente para grafos dispersos.

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float('infinity') for node in graph}  # O(V) - V es el número de vértices
    distances[start] = 0
    pq = [(0, start)]                                        # O(1) - Inicialización del heap

    while pq:                                                # O(E log V) - E es el número de aristas
        current_distance, current_node = heapq.heappop(pq)   # O(log V) - Operación de heap

        if current_distance > distances[current_node]:       # O(1) - Comprobación simple
            continue

        for neighbor, weight in graph[current_node].items(): # O(V) en el peor caso
            distance = current_distance + weight             # O(1) - Suma simple
            if distance < distances[neighbor]:               # O(1) - Comparación simple
                distances[neighbor] = distance               # O(1) - Actualización
                heapq.heappush(pq, (distance, neighbor))     # O(log V) - Operación de heap

    return distances

# Complejidad total: O((V + E) log V)
# Se calcula considerando que cada vértice y arista se procesa una vez,
# y cada operación de heap es O(log V)
```
