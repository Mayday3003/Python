```python
def fibo_memo(n: int, memo = {}):
    if n <= 1:
        return n
    if n in memo:
       return memo[n]

    memo[n] = fibo_memo(n-1) + fibo_memo(n-2) 
    return memo[n]

fibo_memo(100)
```

```python
def coin_change(monedas: list[int], cantidad: int, memo: dict = None) -> int:

    if memo is None:
        memo = {}
    #paso 1 
    if cantidad == 0:
        return 0
    if cantidad < 0:
        return float('inf')
    
    #Paso 2
    if cantidad in memo:
        return memo[cantidad]

    min_monedas = float('inf')
    for coin in monedas:
        result = coin_change(monedas, cantidad - coin, memo)
        min_monedas = min(min_monedas, result + 1)
        print(memo)
        
    #Paso 3
    memo[cantidad] = min_monedas
    return memo[cantidad] if memo[cantidad] != float('inf') else -1





# Ejemplo
monedas = [1, 5, 10, 25]
cantidad = 50

result = coin_change(monedas, cantidad)
print(monedas)
print(cantidad)
print(result)
```
