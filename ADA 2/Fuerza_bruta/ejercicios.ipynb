```python
def suma_Fbruta(lista: list[int], objetivo: int) -> bool:
    n = len(lista)
    for i in range(n):
        for j in range(i + 1, n):
            if lista[i] + lista[j] == objetivo:
                return True, (lista[i], lista[j])
    return False, None


suma_Fbruta([100,50,300,220,10],400)
```

```python
def suma_dos(lista, objetivo):
    visitados = set()
    for i in lista:
        complemento = objetivo - i
        if complemento in visitados:
            return True
        visitados.add(i)
    return False


suma_dos([100,50,300,220,10],400)
```

```python
def factorial(n: int) -> int:
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


factorial(5)  
```

```python
def fibonnaci(n: int):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonnaci(n-1) + fibonnaci(n-2)

fibonnaci(50)



```
